```python
stock_zcfz_em_df_sorted_by_cash_top1000 = stock_zcfz_em_df_sorted_by_cash.head(1000)
```

```python
stock_zcfz_em_df_sorted_by_cash_top1000["特殊指标"] = 0
stock_zcfz_em_df_sorted_by_cash_top1000.head(1)
```

```python
# 遍历2023年第三季报
for index, row in stock_zcfz_em_df_sorted_by_cash_top1000.iterrows():
  stock_zh_valuation_baidu = ak.stock_zh_valuation_baidu(symbol=row['股票代码'], indicator="总市值", period="近一年")
  row['特殊指标'] = row['资产-货币资金'] / (stock_zh_valuation_baidu[stock_zh_valuation_baidu['date'] == pd.to_datetime(curr_date).date()]['value'].values[0]*100000000)
  print(row['股票代码'],"-",row['股票简称'],":",row['特殊指标'])
```
