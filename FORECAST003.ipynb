```python
import yfinance as yf
import pandas as pd

# 设置股票代码
ticker = "0005.HK"

# 获取历史数据
data = yf.download(ticker, period="12mo")

# 扁平化多重索引
#data.columns = [f"{col[0]}_{col[1]}" for col in data.columns]
data.columns = [f"{col[0]}" for col in data.columns]

# 打印调整后的列名
print(data.columns)

# 打印调整后的数据
print(data)
data.to_csv('data.csv')
```

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import numpy as np  # 导入 numpy 库，用于数值计算
import matplotlib.pyplot as plt  # 导入 matplotlib.pyplot 库，用于绘图
import statsmodels.api as sm  # 导入 statsmodels.api 库，用于统计建模
from statsmodels.tsa.stattools import adfuller  # 导入 adfuller 函数，用于 ADF 平稳性检验
from statsmodels.tsa.seasonal import seasonal_decompose  # 导入 seasonal_decompose 函数，用于时间序列分解
from statsmodels.tsa.arima.model import ARIMA  # 导入 ARIMA 模型类
from dateutil.relativedelta import relativedelta  # 导入 relativedelta 类，用于日期计算
import warnings  # 导入 warnings 模块，用于抑制警告
import plotly.graph_objects as go  # 导入 plotly.graph_objects 模块，用于交互式绘图
from plotly.subplots import make_subplots  # 导入 make_subplots 函数，用于创建多子图
import scipy.stats as scs  # 导入 scipy.stats 模块，用于科学统计计算
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf  # 导入 plot_acf 和 plot_pacf 函数，用于绘制 ACF 和 PACF 图
from statsmodels.stats.diagnostic import acorr_ljungbox  # 导入 Ljung-Box 检验函数
from scipy.stats import zscore  # 导入 zscore 函数，用于计算 Z-score


# 1. 读取数据并进行预处理
def load_and_preprocess_data(file_path, sheet_name=None):
    """
    读取CSV或Excel文件，进行日期转换、排序和缺失值处理。

    参数:
    file_path (str): CSV或Excel文件路径。
    sheet_name (str, 可选): 要读取的工作表名称（仅限Excel）。默认为None，即读取第一个工作表。

    返回:
    pandas.DataFrame: 预处理后的数据。
    """
    try:
        if file_path.endswith('.csv'):  # 如果文件路径以 .csv 结尾
            df = pd.read_csv(file_path)  # 使用 pandas 读取 CSV 文件
        elif file_path.endswith(('.xls', '.xlsx')):  # 如果文件路径以 .xls 或 .xlsx 结尾
            df = pd.read_excel(file_path, sheet_name=sheet_name)  # 使用 pandas 读取 Excel 文件
        else:
            raise ValueError("Unsupported file type. Please use .csv, .xls, or .xlsx files.")  # 如果文件类型不支持，则抛出异常

        # 检查 'Date' 列是否存在
        if 'Date' not in df.columns:
            raise KeyError("'Date' column not found in the data file.")

        df["Date"] = pd.to_datetime(df["Date"])  # 将 'Date' 列转换为 datetime 类型
        df = df.sort_values(by="Date")  # 按 'Date' 列排序
        df = df.set_index("Date")  # 将 'Date' 列设置为索引
        df = df.interpolate(method='time')  # 使用时间插值填充缺失值
        return df  # 返回预处理后的数据
    except FileNotFoundError:
        print(f"错误：文件未找到。请检查路径：{file_path}")  # 打印错误信息
        return None  # 返回 None
    except KeyError as e:
        print(f"错误：{e}")  # 打印错误信息
        return None  # 返回 None
    except ValueError as e:
        print(f"错误：{e}")  # 打印错误信息
        return None  # 返回 None
    except Exception as e:
        print(f"读取文件时发生错误：{e}")  # 打印错误信息
        return None  # 返回 None


# 2. 平稳性检验 (ADF检验)
def stability_test(time_series, significance_level=0.05):
    """
    对给定的时间序列进行平稳性检验（ADF检验）。

    参数:
    time_series (pandas.Series): 要检验的时间序列。
    significance_level (float, 可选): 显著性水平。 默认值为0.05。

    返回:
    bool: 如果序列平稳，则返回True，否则返回False。
    """
    try:
        adf_result = adfuller(time_series)  # 进行 ADF 检验
        adf_statistic, p_value, _, _, critical_values, _ = adf_result  # 获取 ADF 检验结果

        print(f'ADF 统计量: {adf_statistic}')  # 打印 ADF 统计量
        print(f'p 值: {p_value}')  # 打印 p 值
        print('临界值:')  # 打印临界值
        for key, value in critical_values.items():  # 遍历临界值字典
            print(f'   {key}: {value}')  # 打印临界值

        if p_value < significance_level:  # 如果 p 值小于显著性水平
            print("序列是平稳的")  # 打印序列平稳
            return True  # 返回 True
        else:
            print("序列是非平稳的")  # 打印序列非平稳
            return False  # 返回 False
    except Exception as e:
        print(f"进行平稳性检验时发生错误：{e}")  # 打印错误信息
        return False  # 返回 False


# 3. 确定差分阶数
def determine_order_of_differencing(data, max_diff=5, significance_level=0.05):
    """
    确定时间序列变为平稳所需的差分阶数。

    参数:
    data (pandas.DataFrame): 包含时间序列数据的数据框。
    max_diff (int, 可选): 最大差分阶数。 默认值为5。
    significance_level (float, 可选): ADF检验的显著性水平。

    返回:
    dict: 一个字典，其中键是列名，值是差分阶数。
          如果某个序列在max_diff阶差分后仍不平稳，则返回None。
    """
    order_of_differencing = {}  # 初始化差分阶数字典
    for col in data.columns:  # 遍历数据框的列
        ts = data[col].dropna()  # 获取当前列的时间序列，并删除缺失值
        diff_order = 0  # 初始化差分阶数为 0
        while diff_order <= max_diff:  # 当差分阶数小于等于最大差分阶数时
            if stability_test(ts, significance_level):  # 如果序列平稳
                order_of_differencing[col] = diff_order  # 记录差分阶数
                break  # 跳出循环
            else:
                ts = ts.diff().dropna()  # 对序列进行差分，并删除缺失值
                diff_order += 1  # 差分阶数加 1
        if diff_order > max_diff:  # 如果差分阶数大于最大差分阶数
            order_of_differencing[col] = None  # 记录差分阶数为 None
            print(f"警告：序列 {col} 在 {max_diff} 阶差分后仍然是非平稳的。")  # 打印警告信息
    return order_of_differencing  # 返回差分阶数字典


# 4. 白噪声检验 (Ljung-Box检验)
def whiteness_test(time_series, lags=20, significance_level=0.05):
    """
    对给定的时间序列进行白噪声检验（Ljung-Box检验）。

    参数:
    time_series (pandas.Series): 要检验的时间序列。
    lags (int, 可选): 检验中使用的滞后阶数。默认值为20。
    significance_level (float, 可选): 显著性水平。

    返回:
    bool: 如果序列是白噪声，则返回True，否则返回False。
    """
    try:
        lb_result = acorr_ljungbox(time_series, lags=lags, boxpierce=False)  # 进行 Ljung-Box 检验
        if lb_result is not None:  # 如果 Ljung-Box 检验结果不为 None
            lb_statistic, p_values = lb_result  # 获取检验统计量和 p 值
            if len(p_values) > 0:
                p_value = p_values[0]
                print(f'Ljung-Box 统计量: {lb_statistic[0]}')  # 打印 Ljung-Box 统计量
                print(f'p 值: {p_value}')  # 打印 p 值
                if p_value < significance_level:  # 如果 p 值小于显著性水平
                    print("序列不是白噪声")  # 打印序列不是白噪声
                    return False  # 返回 False
                else:
                    print("序列是白噪声")  # 打印序列是白噪声
                    return True  # 返回 True
            else:
                print("Ljung-Box 检验返回空的 p 值列表，无法确定是否为白噪声。")  # 打印错误信息
                return False
        else:
            print("Ljung-Box 检验返回 None，无法确定是否为白噪声。")  # 打印错误信息
            return False
    except Exception as e:
        print(f"进行白噪声检验时发生错误：{e}")  # 打印错误信息
        return False  # 返回 False


# 5. 绘制时间序列图
def plot_time_series(data, title="Time Series Data"):
    """
    使用Plotly绘制数据框中的多条时间序列，并将Volume绘制在右侧y轴。

    参数:
    data (pandas.DataFrame): 包含时间序列数据的数据框。
    title (str): 图表的标题。
    """
    try:
        # 检查 'Volume' 列是否存在
        if 'Volume' not in data.columns:
            print("警告: 找不到 'Volume' 列。正在绘制其他序列。")  # 打印警告
            data_to_plot = data.drop(columns=['Volume'], errors='ignore')  # 创建一个不包含 'Volume' 的副本
        else:
            data_to_plot = data.copy()  # 复制一份，不要修改原始数据

        data_melted = data_to_plot.reset_index().melt(id_vars='Date', var_name='Series', value_name='Value')  # 将数据转换为长格式

        fig = make_subplots(specs=[[{"secondary_y": True}]])  # 创建包含双 y 轴的子图

        # 添加其他序列的折线图
        for series in data_melted['Series'].unique():  # 遍历每个序列
            if series != 'Volume':  # 如果序列不是 'Volume'
                fig.add_trace(  # 添加折线图
                    go.Scatter(
                        x=data_melted[data_melted['Series'] == series]['Date'],  # x 轴数据为日期
                        y=data_melted[data_melted['Series'] == series]['Value'],  # y 轴数据为序列值
                        name=series  # 设置序列名称
                    ),
                    secondary_y=False  # 使用左侧 y 轴
                )

        # 添加 Volume 的折线图，并指定y轴
        if 'Volume' in data.columns:
            fig.add_trace(
                go.Scatter(
                    x=data_melted[data_melted['Series'] == 'Volume']['Date'],
                    y=data_melted[data_melted['Series'] == 'Volume']['Value'],
                    name='Volume',
                    marker_color='orange'
                ),
                secondary_y=True,  # 使用右侧 y 轴
            )

        # 设置轴标题
        fig.update_layout(
            title=title,  # 设置图表标题
            xaxis_title='Date',  # 设置 x 轴标题
            yaxis_title='Value',  # 设置左侧 y 轴标题
            yaxis2_title='Volume'  # 设置右侧 y 轴标题
        )

        fig.show()  # 显示图表
    except Exception as e:
        print(f"绘制时间序列图时发生错误：{e}")  # 打印错误信息



# 6. 时间序列分解
def perform_decomposition(time_series, model='additive', period=30):
    """
    对给定的时间序列进行分解。

    参数:
    time_series (pandas.Series): 要分解的时间序列。
    model (str, 可选): 分解模型，'additive' 或 'multiplicative'。 默认值为 'additive'。
    period (int, 可选): 季节性成分的周期。 默认值为30。

    返回:
    statsmodels.tsa.seasonal.DecomposeResult: 分解结果对象。
    """
    try:
        decomposition = seasonal_decompose(time_series, model=model, period=period)  # 进行时间序列分解
        return decomposition  # 返回分解结果
    except Exception as e:
        print(f"进行时间序列分解时发生错误：{e}")  # 打印错误信息
        return None  # 返回 None


# 7. 绘制时间序列分解结果
def plot_decomposition(decomposition, title="Time Series Decomposition"):
    """
    绘制时间序列分解的结果。

    参数:
    decomposition (statsmodels.tsa.seasonal.DecomposeResult): 分解结果对象。
    title (str, 可选): 图表的标题。
    """
    try:
        if decomposition is None:  # 如果分解结果为 None
            print("由于分解失败，无法绘制分解结果。")  # 打印错误信息
            return

        fig, axs = plt.subplots(4, 1, figsize=(12, 8), sharex=True)  # 创建 4 个子图
        axs[0].plot(decomposition.observed, label='Original Data')  # 绘制原始数据
        axs[0].set_title('Original Time Series')  # 设置子图标题
        axs[0].legend(loc='upper left')  # 设置图例

        axs[1].plot(decomposition.trend, label='Trend', color='red')  # 绘制趋势成分
        axs[1].set_title('Trend Component')  # 设置子图标题
        axs[1].legend(loc='upper left')  # 设置图例

        axs[2].plot(decomposition.seasonal, label='Seasonality', color='green')  # 绘制季节性成分
        axs[2].set_title('Seasonal Component')  # 设置子图标题
        axs[2].legend(loc='upper left')  # 设置图例

        axs[3].plot(decomposition.resid, label='Residual', color='purple')  # 绘制残差成分
        axs[3].set_title('Residual Component')  # 设置子图标题
        axs[3].legend(loc='upper left')  # 设置图例

        plt.suptitle(title, fontsize=16)  # 设置总标题
        plt.tight_layout()  # 调整子图布局
        plt.show()  # 显示图表
    except Exception as e:
        print(f"绘制时间序列分解结果时发生错误：{e}")  # 打印错误信息



# 8. 绘制自相关和偏自相关函数
def plot_acf_pacf(time_series, lags=20, title="ACF and PACF"):
    """
    绘制给定时间序列的自相关函数 (ACF) 和偏自相关函数 (PACF)。

    参数:
    time_series (pandas.Series): 要分析的时间序列。
    lags (int, 可选): 要绘制的滞后数。 默认值为20。
    title (str, 可选): 图表的标题。
    """
    try:
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))  # 创建 2 个子图
        plot_acf(time_series, lags=lags, ax=ax1, alpha=0.05)  # 绘制 ACF 图
        ax1.set_title('Autocorrelation Function (ACF)')  # 设置子图标题
        plot_pacf(time_series, lags=lags, ax=ax2, alpha=0.05)  # 绘制 PACF 图
        ax2.set_title('Partial Autocorrelation Function (PACF)')  # 设置子图标题
        plt.suptitle(title, fontsize=16)  # 设置总标题
        plt.tight_layout()  # 调整子图布局
        plt.show()  # 显示图表
    except Exception as e:
        print(f"绘制 ACF 和 PACF 图时发生错误：{e}")  # 打印错误信息



# 9. 新增的时间序列可视化分析函数
def ts_plot(y, lags=None, title='Time Series Analysis Plots', figsize=(12, 10)):
    """
    绘制时间序列的多个分析图，包括原始序列、ACF、PACF、QQ图和PP图。

    参数:
    y (pandas.Series): 要分析的时间序列。
    lags (int, 可选): ACF和PACF图中要显示的滞后数。 默认值为None。
    title (str, 可选): 图表的总标题。 默认值为 'Time Series Analysis Plots'。
    figsize (tuple, 可选): 图表的大小。 默认值为 (12, 10)。
    """
    if not isinstance(y, pd.Series):  # 如果 y 不是 pandas.Series 类型
        y = pd.Series(y)  # 将 y 转换为 pandas.Series 类型
    fig = plt.figure(figsize=figsize)  # 创建图表
    layout = (3, 2)  # 设置子图布局
    ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)  # 创建原始时间序列子图
    acf_ax = plt.subplot2grid(layout, (1, 0))  # 创建 ACF 子图
    pacf_ax = plt.subplot2grid(layout, (1, 1))  # 创建 PACF 子图
    qq_ax = plt.subplot2grid(layout, (2, 0))  # 创建 QQ 图子图
    pp_ax = plt.subplot2grid(layout, (2, 1))  # 创建 PP 图子图

    y.plot(ax=ts_ax)  # 绘制原始时间序列
    ts_ax.set_title('Time Series Analysis')  # 设置子图标题
    plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)  # 绘制 ACF 图
    plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)  # 绘制 PACF 图
    sm.qqplot(y, line='s', ax=qq_ax)  # 绘制 QQ 图
    qq_ax.set_title('QQ Plot')  # 设置子图标题
    scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)  # 绘制 PP 图
    pp_ax.set_title('PP Plot')  # 设置子图标题

    plt.suptitle(title, fontsize=16)  # 设置总标题
    plt.tight_layout()  # 调整子图布局
    plt.show()  # 显示图表



# 10. ARIMA模型类
class arima_model:
    def __init__(self, ts, p=1, d=0, q=1):
        """
        初始化ARIMA模型类。

        参数:
        ts (pandas.Series): 输入的时间序列数据。
        p (int, 可选): AR模型的阶数。 默认值为1。
        d (int, 可选): 差分阶数。 默认值为0。
        q (int, 可选): MA模型的阶数。 默认值为1。
        """
        self.data_ts = ts  # 存储时间序列数据
        self.p = p  # 存储 AR 阶数
        self.d = d  # 存储差分阶数
        self.q = q  # 存储 MA 阶数
        self.model = None  # 初始化 ARIMA 模型
        self.results = None  # 初始化模型拟合结果

    def fit(self):
        """
        拟合ARIMA模型。
        """
        try:
            self.model = ARIMA(self.data_ts, order=(self.p, self.d, self.q))  # 创建 ARIMA 模型实例
            self.results = self.model.fit()  # 拟合模型
        except Exception as e:
            print(f"拟合 ARIMA 模型时发生错误：{e}")  # 打印错误信息
            self.results = None  # 将拟合结果设为 None

    def forecast(self, steps=1):
        """
        预测未来的值。

        参数:
        steps (int, 可选): 要预测的步数。 默认值为1。

        返回:
        pandas.Series: 预测值。
        """
        if self.results is None:  # 如果模型未拟合
            print("模型未拟合，无法进行预测。")  # 打印错误信息
            return None  # 返回 None
        try:
            forecast_result = self.results.forecast(steps=steps)  # 进行预测
            return forecast_result  # 返回预测结果
        except Exception as e:
            print(f"进行预测时发生错误：{e}")  # 打印错误信息
            return None  # 返回 None

    def plot_results(self, title="ARIMA Model Results"):
        """
        绘制原始数据和模型的预测结果。

        参数:
        title (str, 可选): 图表的标题。
        """
        if self.results is None:  # 如果模型未拟合
            print("模型未拟合，无法绘制结果。")  # 打印错误信息
            return
        try:
            plt.figure(figsize=(12, 6))  # 创建图表
            plt.plot(self.data_ts.index, self.data_ts, label='Original Data', color='blue')  # 绘制原始数据
            plt.plot(self.data_ts.index, self.results.fittedvalues, label='Fitted Values', color='red')  # 绘制拟合值
            forecast_index = pd.to_datetime(self.data_ts.index[-1]) + pd.to_timedelta(np.arange(1, 11), 'D')
            plt.plot(forecast_index, self.forecast(steps=10), label='Forecasted Values', color='green')  # 绘制预测值

            plt.title(title)  # 设置图表标题
            plt.xlabel('Date')  # 设置 x 轴标签
            plt.ylabel('Value')  # 设置 y 轴标签
            plt.legend()  # 显示图例
            plt.grid()  # 显示网格线
            plt.show()  # 显示图表
        except Exception as e:
            print(f"绘制模型结果时发生错误：{e}")  # 打印错误信息




if __name__ == '__main__':
    # 1. 读取数据
    file_path = 'data.csv'
    df = load_and_preprocess_data(file_path)
    if df is None:
        exit()

    # 2. 绘制原始时间序列
    plot_time_series(df, title="Original Time Series Data")

    # 3. 进行平稳性检验并确定差分阶数
    print("\nStationarity Test and Determining Order of Differencing:")
    orders = determine_order_of_differencing(df)
    print("Recommended Order of Differencing:", orders)

    # 4. 对平稳序列进行建模和预测
    for col in df.columns:
        print(f"\nARIMA Modeling and Forecasting for column '{col}':")
        ts = df[col]
        # 假设我们已经通过其他方式确定了p, d, q
        d = orders[col]  # 使用之前确定的差分阶数
        p, q = 1, 1  # 实际应用中需要通过AIC、BIC等准则确定
        model = arima_model(ts, p=p, d=d, q=q)
        model.fit()
        if model.results is not None:
            model.plot_results(title=f"{col} - ARIMA Model Results")
            forecast_value = model.forecast(steps=1)
            print(f"Predicted next value ({col}): {forecast_value.iloc[0]:.2f}")
            # 5. 时间序列可视化分析
            print(f"\nTime Series Analysis for column '{col}':")
            ts_plot(ts, lags=20, title=f"{col} Time Series Analysis")
        else:
            print(f"ARIMA model fitting failed for column '{col}'.")

    # 对原始数据 df 也进行时间序列分解和异常点检测
    print("\nTime Series Decomposition and Anomaly Detection for Original Data:")
    for col in df.columns:
        print(f"\nTime Series Decomposition for column '{col}':")
        ts = df[col]
        decomposition = perform_decomposition(ts, model="additive", period=30)
        if decomposition is not None:
            plot_decomposition(decomposition, title=f"{col} Time Series Decomposition")

            # 异常点检测与绘制
            print(f"\nAnomaly Detection and Plotting for column '{col}':")
            # 计算残差的 Z-score
            residuals = decomposition.resid
            if residuals is not None:
                z_scores = zscore(residuals.dropna())
                anomalies = np.abs(z_scores) > 2.5

                # 绘制异常点
                plt.figure(figsize=(12, 6))
                plt.plot(ts.index, ts, label=f"{col}", color="blue")
                anomaly_dates = ts.index[anomalies.reindex(ts.index, fill_value=False)]  # 获取异常点的日期
                plt.scatter(anomaly_dates, ts.loc[anomaly_dates], color="red", label="Anomalies", marker="o")
                plt.title(f"{col} with Anomalies")
                plt.xlabel("Date")
                plt.ylabel("Value")
                plt.legend()
                plt.grid()
                plt.show()
            else:
                print(f"Decomposition residuals are None, cannot perform anomaly detection for column '{col}'.")
        else:
            print(f"Time series decomposition failed for column '{col}'.")
```

```python
import numpy as np  # 导入 numpy 库，用于数值计算
import matplotlib.pyplot as plt  # 导入 matplotlib.pyplot 库，用于绘图

# 生成时间序列 (例如，20 年的每月数据)
np.random.seed(42)  # 设置随机数种子，确保结果可重复
time = np.arange(0, 240, 1)  # 240 个月 (20 年)

# 趋势：二次增长后下降
trend = 0.02 * (time - 120) ** 2 - 50

# 季节性：使用正弦波表示的年度周期
seasonality = 5 * np.sin(2 * np.pi * time / 12)

# 周期性成分：长期经济周期（约 5 年）
cycle = 3 * np.sin(2 * np.pi * time / 60)

# 随机性：使用高斯噪声表示市场波动
randomness = np.random.normal(scale=3, size=len(time))

# 组合所有成分
simulated_stock_price = trend + seasonality + cycle + randomness

# 绘制模拟股价
plt.figure(figsize=(12, 5))  # 创建一个 12x5 英寸的图
plt.plot(time, simulated_stock_price, label="Simulated Stock Price", color='b')  # 绘制模拟股价，蓝色实线
plt.plot(time, trend, label="Trend (T)", linestyle="dashed", color='r')  # 绘制趋势，红色虚线
plt.plot(time, seasonality, label="Seasonality (S)", linestyle="dotted", color='g')  # 绘制季节性，绿色点线
plt.plot(time, cycle, label="Cycle (C)", linestyle="dashdot", color='orange')  # 绘制周期性，橙色点划线
plt.legend()  # 显示图例
plt.title("Simulated Stock Price with Trend, Seasonality, and Cycles")  # 设置图标题
plt.xlabel("Time (Months)")  # 设置 x 轴标签
plt.ylabel("Stock Price")  # 设置 y 轴标签
plt.show()  # 显示绘图
print("绘图完成")
```
