### 乌龟比赛 画法

```python
from ColabTurtlePlus.Turtle import *
import random
from time import sleep
```

```python
##设置 窗口
clearscreen()
setup(800,500)
bgcolor("forestgreen")
speed(0)

##绘制 顶部
#抬起笔头 移动不会生成绘画
penup()
#放下笔头 移动就会生成绘画
#pendown()
#移动笔头 落在 目标位置
goto(0,215)
#设置笔头 颜色
color("white")
#输出文字
write("TURTLE RACE", align="center", font=("arial", 20, "bold"))



##绘制 泥土
#移动笔头 落在 目标位置
goto(-350,200)
#放下笔头 移动就会生成绘画
pendown()
#设置笔头 颜色
color("saddlebrown")
#开始填充
begin_fill()
for i in range(2):
  forward(700)
  right(90)
  forward(400)
  right(90)
#结束填充
end_fill()

##绘制 终线
gap_size = 20
##设置笔头 形状
shape("square")
#抬起笔头 移动不会生成绘画
penup()
#设置笔头 颜色
color("white")
for i in range(10):
  goto(250,(170-(i*gap_size*2)))
  #用笔头形状盖个章 1代表图层上 0代表图层下 
  stamp(layer=1)
for i in range(10):
  goto(250+gap_size,(210-gap_size-(i*gap_size*2)))
  #用笔头形状盖个章 1代表图层上 0代表图层下
  stamp(layer=1)
#设置笔头 颜色
color("black")
for i in range(10):
  goto(250,(190-(i*gap_size*2)))
  #用笔头形状盖个章 1代表图层上 0代表图层下
  stamp(layer=1)
for i in range(10):
  goto(251+gap_size,(190-gap_size-(i*gap_size*2)))
  #用笔头形状盖个章 1代表图层上 0代表图层下
  stamp(layer=1)
done()

##创建 乌龟 奔跑
#创建 四支画笔
turtles = [Turtle(), Turtle(), Turtle(), Turtle()]
for i in range(4): turtles[i].hideturtle()
colors = ["cyan", "magenta", "yellow", "lime"]
for i in range(4):
  #设置 笔头成为乌龟形状
  turtles[i].shape("turtle2")
  #设置 形状颜色
  turtles[i].color(colors[i])
  #设置 形状大小
  turtles[i].shapesize(1.5)
  #设置 笔头速度
  turtles[i].speed(10)
  #设置 显示形状
  turtles[i].showturtle()
  turtles[i].penup()
  turtles[i].goto(-300,150-100*i)
  turtles[i].pendown()

##赛前 暂停 一秒
time.sleep(1)

##开始比赛
random.seed()
#xcor() 获得 当前画笔X坐标
#ycor() 获得 当前画笔Y坐标
positions = [turtles[i].xcor() for i in range(4)]
while max(positions) <= 230:
  turtles[0].forward(random.randint(1,10))
  turtles[1].forward(random.randint(1,10))
  turtles[2].forward(random.randint(1,10))
  turtles[3].forward(random.randint(1,10))
  positions = [turtles[i].xcor() for i in range(4)]

##查看谁是赢家
win_pos = max(positions)
winners = [i for i, j in enumerate(positions) if j == win_pos]
for k in winners:
  turtles[k].speed(5)
  turtles[k].left(360)

##打印谁是赢家
hideturtle()
color("white")
goto(0, -235)
if len(winners)==1:
  write("Turtle "+colors[winners[0]]+" won!", align="center", font=("Arial", 20, "bold"))
else:
  s = colors[winners[0]]
  for k in range(1,len(winners)):
    s += " and " + colors[winners[k]]
  write("Turtles "+s+ " tied!", align="center", font=("Arial", 20, "bold"))
done()

##打印最后位置
pad = max([len(s) for s in colors])+10
for k in range(4):
  print(("turtle "+colors[k]).ljust(pad), int(positions[k]))
```

### 向日葵 画法

```python
from ColabTurtlePlus.Turtle import *
```

```python
import math	

def drawPhyllPattern(turtle, numseeds, numpetals, angle, cspread):
	"""print a pattern of circles using spiral phyllotactic data"""
	# 设置 初始位置
	turtle.color('black')
	turtle.fillcolor("orange")
	phi = angle * ( math.pi / 180.0 ) 
	xcenter = 0.0
	ycenter = 0.0
	#xcenter = ( 0.0 + window_width() ) / 2
	#ycenter = ( 0.0 + window_height() ) / 2
	
	for n in range(numseeds+numpetals):
		r = cspread * math.sqrt(n)
		theta = n * phi
		
		x = r * math.cos(theta) + xcenter
		y = r * math.sin(theta) + ycenter

		# 移动画笔到指定位置 （移动过程不留笔迹）
		turtle.up()
		turtle.setpos(x, y)
		turtle.down()
		# 调整画笔朝向
		turtle.setheading(n * angle)
		if n < numseeds:
			turtle.stamp()
		else:
			drawPetal(turtle)

def drawPetal(turtle):
	turtle.color('black')
	turtle.fillcolor('yellow')
	turtle.begin_fill()
	turtle.right(20)
	turtle.forward(70)
	turtle.left(40)
	turtle.forward(70)
	turtle.left(140)
	turtle.forward(70)
	turtle.left(40)
	turtle.forward(70)
	turtle.end_fill()
	turtle.penup()
```

```python
clearscreen()
setup(450,450)
bgcolor('green')
gfg = Turtle()
#speed(0) 直接展示最终结果 不用显示中间动画
gfg.speed(50)
gfg.shape("turtle2")
drawPhyllPattern(gfg, 165, 35, 137.508, 5)
gfg.ht()
gfg.done()
saveSVG("phyllotaxis")
```

### 阴阳图 画法

```python
from ColabTurtlePlus.Turtle import *
```

```python
def yinyan(radius, color1, color2):
    width(3)
    pencolor("black")
    fillcolor(color1)
    begin_fill()
    circle(radius/2., 180)
    circle(radius, 180)
    left(180)
    circle(-radius/2., 180)
    end_fill()
    fillcolor(color2)
    left(90)
    up()
    forward(radius*0.375)
    right(90)
    down()
    begin_fill()
    circle(radius*0.125)
    left(90)
    end_fill()
    up()
    backward(radius*0.375)
    down()
    left(90)
```

```python
clearscreen()
bgcolor('green')
yinyan(200, "white", "black")
yinyan(200, "black", "white")
ht()
```

```python
from ColabTurtlePlus.Turtle import *
```

```python
def design(homePos, scale):
    t.up()
    for i in range(5):
        t.forward(64.65 * scale)
        t.down()
        wheel(t.position(), scale)
        t.up()
        t.backward(64.65 * scale)
        t.right(72)
    t.up()
    t.goto(homePos)
    t.right(36)
    t.forward(24.5 * scale)
    t.right(198)
    t.down()
    centerpiece(46 * scale, 143.4, scale)
    t.done()

def wheel(initpos, scale):
    t.right(54)
    for i in range(4):
        pentpiece(initpos, scale)
    t.down()
    t.left(36)
    for i in range(5):
        tripiece(initpos, scale)
    t.left(36)
    for i in range(5):
        t.down()
        t.right(72)
        t.forward(28 * scale)
        t.up()
        t.backward(28 * scale)
    t.left(54)
    t.done()

def tripiece(initpos, scale):
    oldh = t.heading()
    t.down()
    t.backward(2.5 * scale)
    tripolyr(31.5 * scale, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.down()
    t.backward(2.5 * scale)
    tripolyl(31.5 * scale, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.left(72)
    t.done()

def pentpiece(initpos, scale):
    oldh = t.heading()
    t.up()
    t.forward(29 * scale)
    t.down()
    for i in range(5):
        t.forward(18 * scale)
        t.right(72)
    pentr(18 * scale, 75, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.forward(29 * scale)
    t.down()
    for i in range(5):
        t.forward(18 * scale)
        t.right(72)
    pentl(18 * scale, 75, scale)
    t.up()
    t.goto(initpos)
    t.setheading(oldh)
    t.left(72)
    t.done()

def pentl(side, ang, scale):
    if side < (2 * scale): return
    t.forward(side)
    t.left(ang)
    pentl(side - (.38 * scale), ang, scale)

def pentr(side, ang, scale):
    if side < (2 * scale): return
    t.forward(side)
    t.right(ang)
    pentr(side - (.38 * scale), ang, scale)

def tripolyr(side, scale):
    if side < (4 * scale): return
    t.forward(side)
    t.right(111)
    t.forward(side / 1.78)
    t.right(111)
    t.forward(side / 1.3)
    t.right(146)
    tripolyr(side * .75, scale)

def tripolyl(side, scale):
    if side < (4 * scale): return
    t.forward(side)
    t.left(111)
    t.forward(side / 1.78)
    t.left(111)
    t.forward(side / 1.3)
    t.left(146)
    tripolyl(side * .75, scale)

def centerpiece(s, a, scale):
    t.forward(s); t.left(a)
    if s < (7.5 * scale):
        return
    centerpiece(s - (1.2 * scale), a, scale)
```

```python
clearscreen()
setup(550,550)
showborder()
t = Turtle()
t.animationOff()
t.speed(0)
t.hideturtle()
design(t.position(), 2)
t.done()
saveSVG("bytedesign")
```
