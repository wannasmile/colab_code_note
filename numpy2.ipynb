# **打印数组**

```
import numpy as np
a = np.arange(6) # 1d array
print(a)
b = np.arange(12).reshape(4,3) # 2d array
print(b)
c = np.arange(24).reshape(2,3,4) # 3d array
print(c)

print(np.arange(10000))
print(np.arange(10000).reshape(100,100))

#np.set_printoptions(threshold=sys.maxsize)
```

当打印数组时，NumPy以与嵌套列表类似的方式显示它，通过以下布局显示：

(d1,d2,...,dm,dn)

> 最后一个轴(dn)从左到右打印；倒数第二个(dm)从上到下打印；其余部分也从上到下打印，每个切片用空行分隔。

> 如果数组太大而无法打印，NumPy会自动跳过数组的中心部分并仅打印角点。

> 要禁用此行为并强制NumPy打印整个数组，可以更改打印选项set_printoptions（np.set_printoptions(threshold=sys.maxsize)）。




# **数学运算**

```
import numpy as np
a = np.array( [20,30,40,50] )
b = np.arange( 4 )
print( b )
c = a-b
print( c )
print( b**2 )
print( 10*np.sin(a) )
print( a<35 )


A = np.array( [[1,1],
               [0,1]] )
B = np.array( [[2,0],
               [3,4]] )

print( A * B ) # elementwise product

print( A @ B ) # matrix product

print( A.dot(B) ) # another matrix product



a = np.ones((2,3), dtype=int)
b = np.random.random((2,3))
a *= 3
b += a

#UFuncTypeError: Cannot cast ufunc 'add' output from dtype('float64') to dtype('int64') with casting rule 'same_kind'
#b is not automatically converted to integer type
#a += b 


a = np.random.random((2,3))
print( a )
print(a.sum())
print(a.min())
print(a.max())



b = np.arange(12).reshape(3,4)
print( b )
print( b.sum(axis=0) ) #shape=(3,4) from left to right: axis 0 # sum of each axis, along axis 0
print( b.min(axis=1) ) #shape=(3,4) from left to right: axis 1 # min of each axis, along axis 1
print( b.cumsum(axis=1) ) #cumulative sum along axis 1
```

数组上的算术运算符会应用到 元素 级别。

与许多矩阵语言不同，乘积运算符*在NumPy数组中按元素进行运算。矩阵乘积可以使用@运算符（在python>=3.5中）或dot函数或方法执行。

某些操作（例如+=和*=）会更直接更改被操作的矩阵数组而不会创建新矩阵数组。

当使用不同类型的数组进行操作时，结果数组的元素类型为更一般类型的数组（称为向上类型转换的行为）。

许多一元操作，例如计算数组中所有元素的总和，都是作为ndarray类的方法实现的。
默认情况下，这些一元操作适用于整个数组，就像它是一个数字列表一样，无论其形状如何。但是，通过指定axis参数，就可以沿数组的指定轴应用操作。
