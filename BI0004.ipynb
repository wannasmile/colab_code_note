```python
import altair as alt
from vega_datasets import data

source = data.cars()

source.head(10)
```

```python
# Configure common options.
# We specify the aggregation as a transform here so we can reuse it in both layers.
base = alt.Chart(source).transform_aggregate(
    mean_horsepower='mean(Horsepower)',
    groupby=['Origin', 'Cylinders']
).encode(
    alt.X('Cylinders:O'),
    alt.Y('Origin:O'),
)

# Configure heatmap
heatmap = base.mark_rect().encode(
    color=alt.Color('mean_horsepower:Q',
        scale=alt.Scale(scheme='viridis'),
        legend=alt.Legend(title="Mean of Horsepower"),
    )
)

# Configure text
text = base.mark_text(baseline='middle').encode(
    text=alt.Text('mean_horsepower:Q', format=".0f"),
    color=alt.condition(
        alt.datum.mean_horsepower > 150,
        alt.value('black'),
        alt.value('white')
    )
)

# Draw the chart
heatmap + text
```

```python
#select 产业,活跃度,count(*) cnt from t group by 产业,活跃度
rs3 = df.groupby(['产业','活跃度']).size().to_frame('cnt').reset_index()
rs3.head(10)

```

```python
# Configure common options.
# We specify the aggregation as a transform here so we can reuse it in both layers.
base = alt.Chart(rs3).transform_aggregate(
    sum_cnt='sum(cnt)',
    groupby=['活跃度', '产业']
).encode(
    alt.X('活跃度:O'),
    alt.Y('产业:O'),
)

# Configure heatmap
heatmap = base.mark_rect().encode(
    color=alt.Color('sum_cnt:Q',
        scale=alt.Scale(scheme='viridis'),
        legend=alt.Legend(title="企业数量"),
    )
)

# Configure text
text = base.mark_text(baseline='middle').encode(
    text=alt.Text('sum_cnt:Q', format=".0f"),
    color=alt.condition(
        alt.datum.sum_cnt > 150,
        alt.value('black'),
        alt.value('white')
    )
)

# Draw the chart
(heatmap + text).properties(width=600,height=400)
```

```python
#select 产业,活跃度,区县,count(*) cnt from t group by 产业,活跃度,区县
rs4 = df.groupby(['产业','活跃度','区县']).size().to_frame('cnt').reset_index()
rs4.head(10)

```

```python
rs4['产业活跃度']=rs4['产业']+rs4['活跃度']
rs4.head(10)
```

```python
# Configure common options.
# We specify the aggregation as a transform here so we can reuse it in both layers.
base = alt.Chart(rs4).transform_aggregate(
    sum_cnt='sum(cnt)',
    groupby=['产业活跃度', '区县']
).encode(
    alt.X('产业活跃度:O'),
    alt.Y('区县:O'),
)

# Configure heatmap
heatmap = base.mark_rect().encode(
    color=alt.Color('sum_cnt:Q',
        scale=alt.Scale(scheme='plasma'),#turbo viridis inferno magma plasma cividis
        legend=alt.Legend(title="企业数量"),
    )
)

# Configure text
text = base.mark_text(baseline='middle').encode(
    text=alt.Text('sum_cnt:Q', format=".0f"),
    color=alt.condition(
        alt.datum.sum_cnt > 2000,
        alt.value('black'),
        alt.value('white')
    )
)

# Draw the chart
(heatmap + text).properties(width=600,height=400)
```

```python
#s = pd.Series([1, 2, 3, 4, 5])
action1 = pd.Series(['a', 'b', 'c', 'd', 'e'])
action2 = pd.Series(['A', 'B', 'C', 'D', 'E'])

df = pd.DataFrame({"单号": s, "操作时间": s ** 2, "操作动作": action1})
df
```

```python
df = pd.concat([pd.DataFrame({"单号": s, "操作时间": s + 1, "操作动作": action1}), pd.DataFrame({"单号": s, "操作时间": s ** 2, "操作动作": action2})])
df
```

```python
#1. 按照单号和操作时间排序
df.sort_values(['单号','操作时间'], inplace=True)

#2. 给每行操作唯一顺序编号
df['id']=range(len(df))
df
```

```python
#3. 实现窗口函数：按照单号聚合，按照操作时间排序，然后向前移动窗口
df.groupby('单号', group_keys=True).apply(lambda x: x.sort_values('操作时间').shift(-1))
```

```python
#3. 实现窗口函数：按照单号聚合，按照操作时间排序，然后向前移动窗口
x = df.groupby('单号', group_keys=True).apply(lambda x: x.sort_values('操作时间').shift(-1))

#4. 得到字典
x.index = x.id-1
x
```

```python
#4. 得到字典：单号与下个操作动作的映射关系
x['操作动作'].dropna().to_dict()
```

```python
#5. 通过字典将每个单号关联到下一个操作动作
dict = x['操作动作'].dropna().to_dict()
df['下个操作动作'] = df['id'].map(dict)
df
```
