```python
#print(list(itertools.product([0,1],repeat=2)))
#print(list(itertools.product([0,1],repeat=3)))

#print(list(filter(lambda a:a[0]==1,[(0,'a'),(1,'b'),(0,'c'),(1,'d')])))
#print(list(map(lambda b:b[1],[(1,'b'),(1,'d')])))
#print(list(map(lambda b:b[1],filter(lambda a:a[0]==1,[(0,'a'),(1,'b'),(0,'c'),(1,'d')]))))

#print(list(zip([0,0,1],['a','b','c'])))

#print(len([]))

#print(list(filter(lambda a:'b' in a,map(lambda b:b[1],filter(lambda a:a[0]==1,[(0,'a'),(1,'b'),(0,'c'),(1,'d')])))))
#print('a'.split(','))

import itertools



def generate_cubelist(cubeList, fixedList=None, levelList=None, jointList=None):

    def map_list(x):
        #print(list(x))
        res = filter(lambda a:a[0]==1,x)
        res = map(lambda b:b[1],res)
        #print(list(res))
        return res
    
    cubeList = cubeList.split(',')

    if fixedList is not None:
        fixedList = fixedList.split(',')
    if levelList is not None:
        levelList = levelList.split('||')
    if jointList is not None:
        jointList = jointList.split('||')
    cubeResult = map(map_list,[zip(j,cubeList) for j in [i for i in itertools.product([0,1],repeat=len(cubeList))]])


    cubeResult_eval = list()#[]
    for i in cubeResult:
        cubeResult_eval.append(list(i))

    cubeResult_eval = cubeResult_eval[1:]
    #    print(list(i))
    #    print(*i)
    #for i in cubeResult:
    #    print(list(i))
    #    print(*i)
    #print(cubeResult_eval)

    #在Python 3中，因为map函数处理的结果是迭代器，而且只能迭代一次，如果需要多次使用，需要提前保存。
    #在Python 2中，则可以多次调用。
    if fixedList is not None:
        for f_ele in fixedList:
            #cubeResult = filter(lambda c:f_ele in c,cubeResult)
            cubeResult = filter(lambda c:f_ele in c,cubeResult_eval)
    if levelList is not None and len(levelList) >=1:
        for lev_ele in levelList:
            lev_ele = lev_ele.split(',')
            if len(lev_ele) > 1:
                for i in range(len(lev_ele)):
                    if i == 0:
                        continue
                    else:
                        cubeResult = filter(lambda a : lev_ele[i] not in a or set([lev_ele[j] for j in range(i+1)]) < set(a),cubeResult)
    if jointList is not None and len(jointList) >=1:
        for joint_ele in jointList:
            joint_ele = joint_ele.split(',')
            if len(joint_ele) > 1:
                cubeResult = filter(lambda a:set(a).intersection(set(joint_ele)) == set([]) or set(joint_ele) < set(a), cubeResult)

    return cubeResult




#参数解释：
#cubeList：要打cube的所有维度，有多少写多少
#fixedList：cube中的固定维度，group by中必须要出现的
#levelList：维度中有层级关系的，例如'vip_level1,vip_level2'，即下一层级维度不能独立上一层级维度存在，如果有多组层级维度，用“||”隔开，例如'vip_level1,vip_level2||fst_org_id,sec_org_id,thrd_org_id'
#jointList：cube维度中，两个或两个以上必须同时存在的打包维度，例如'id1,name1'，如果有多个组合，用“||”隔开，例如'id1,name1||id2,name2'
#备注：如果fixedList、levelList、jointList为空，则生成一个cubeList全维度的cube，打cube时运行速度缓慢，fixedList、levelList、jointList是对cube进行裁剪

#交易11个维度cube例子：
for i in generate_cubelist('pay_type,vip_lvl1,vip_lvl2,lst_vip_type1,lst_vip_type2,prod_type,prod_atrrib,prod_pkg,fst_org_id,sec_org_id,thrd_org_id',fixedList='fst_org_id',levelList='vip_lvl1,vip_lvl2||lst_vip_type1,lst_vip_type2||fst_org_id,sec_org_id,thrd_org_id||prod_type,prod_atrrib',jointList='sec_org_id,thrd_org_id'):
    print(list(i))
```
