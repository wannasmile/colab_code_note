What is Correlation?


P(A): Probability that event A occurs \\
P(A’): Probability that event A does not occur \\
P(AB): Probability that events A and B occur together \\

Events A and B are said to be independent if P(AB) = P(A) x P(B). 
Otherwise A and B are dependent.

Events A and B are said to be correlated if any of AB, A’B, AB’, A’B’ are dependent.


A correlation rule is a set of items that are correlated.
