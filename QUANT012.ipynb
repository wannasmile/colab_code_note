```python
from IPython import get_ipython # 导入 IPython 库，用于与 IPython 内核交互
from IPython.display import display # 导入 display 函数，用于在输出中显示内容

# %%
!wget https://raw.githubusercontent.com/StellarCN/scp_zh/refs/heads/master/fonts/SimHei.ttf # 下载 SimHei 字体文件
!fc-list | grep simhei  # 验证字体是否安装成功并获取字体路径

import matplotlib.pyplot as plt # 导入 matplotlib.pyplot 库，用于绘图
import matplotlib.font_manager as fm # 导入 matplotlib.font_manager 库，用于管理字体
from datetime import date, datetime # 导入 date 和 datetime 类，用于处理日期和时间

# 更新 font_path 变量，根据 fc-list 命令的输出设置字体路径
font_path = 'SimHei.ttf'
fontprop = fm.FontProperties(fname=font_path, family='SimHei') # 创建 FontProperties 对象，指定字体文件和字体名称

# 刷新字体缓存
fm.fontManager.addfont(font_path)

# 将 SimHei 设置为整个图表的默认字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 用于正确显示负号

import numpy as np # 导入 numpy 库，用于数值计算
import pandas as pd # 导入 pandas 库，用于数据处理
from datetime import date # 导入 date 类，用于处理日期
from scipy import stats, signal # 导入 scipy 库中的 stats 和 signal 模块，用于统计分析和信号处理
from tqdm import tqdm # 导入 tqdm 库，用于显示进度条

class SupportPressureTester:
    def __init__(self, min15_path, min5_path, window_sizes=[20, 40, 60]):
        """
        初始化测试器

        参数：
            min15_path (str): 15分钟数据文件的路径。
            min5_path (str): 5分钟数据文件的路径。
            window_sizes (list): 回测窗口大小（以小时为单位）。
        """
        # 读取分时数据
        self.df_15min = pd.read_csv(min15_path)  # 读取15分钟数据
        self.df_5min = pd.read_csv(min5_path)  # 读取5分钟数据

        # 确保日期格式正确
        self.df_15min['date'] = pd.to_datetime(self.df_15min['date'])  # 将15分钟数据的日期列转换为日期时间格式
        self.df_5min['date'] = pd.to_datetime(self.df_5min['date'])  # 将5分钟数据的日期列转换为日期时间格式

        # 获取当前日期
        current_date = pd.Timestamp.today()  # 获取当前日期

        # 过滤未来数据
        self.df_15min = self.df_15min[self.df_15min['date'] <= current_date]  # 过滤15分钟数据中日期大于当前日期的数据
        self.df_5min = self.df_5min[self.df_5min['date'] <= current_date]  # 过滤5分钟数据中日期大于当前日期的数据

        # 按日期排序
        self.df_15min = self.df_15min.sort_values('date').reset_index(drop=True)  # 按日期对15分钟数据进行排序，并重置索引
        self.df_5min = self.df_5min.sort_values('date').reset_index(drop=True)  # 按日期对5分钟数据进行排序，并重置索引

        # 获取数据范围
        self.start_date = self.df_15min['date'].min()  # 获取数据的起始日期
        self.end_date = self.df_15min['date'].max()  # 获取数据的结束日期

        print(f"数据范围: {self.start_date.strftime('%Y-%m-%d %H:%M')} 至 {self.end_date.strftime('%Y-%m-%d %H:%M')}")  # 打印数据范围

        self.window_sizes = window_sizes  # 回测窗口大小（以小时为单位）
        self.results = []  # 初始化结果列表

    def calculate_levels(self, data_15min, data_5min):
        """
        计算支撑压力位

        参数：
            data_15min (pd.DataFrame): 15分钟训练数据。
            data_5min (pd.DataFrame): 5分钟训练数据。

        返回：
            dict: 包含支撑位、压力位和趋势信息的字典。
        """
        if len(data_15min) == 0 or len(data_5min) == 0:
            return {'支撑位': {}, '压力位': {}, '趋势': 'unknown'}  # 如果数据为空，返回空结果

        # 使用15分钟数据计算主要水平
        highs_15 = data_15min['high'].values  # 获取15分钟数据的最高价
        lows_15 = data_15min['low'].values  # 获取15分钟数据的最低价
        closes_15 = data_15min['close'].values  # 获取15分钟数据的收盘价
        volumes_15 = data_15min['volume'].values  # 获取15分钟数据的成交量
        current_price = closes_15[-1]  # 获取当前价格（最后一个收盘价）

        # 计算价格分布
        all_prices = np.concatenate([lows_15, highs_15, closes_15])  # 将最高价、最低价和收盘价合并
        kde = stats.gaussian_kde(all_prices)  # 使用高斯核密度估计计算价格分布
        price_range = np.linspace(min(all_prices), max(all_prices), 100)  # 创建价格范围
        density = kde(price_range)  # 计算价格范围内的密度

        # 寻找密集交易区域
        peaks, _ = signal.find_peaks(density, distance=10)  # 寻找密度峰值，作为潜在的支撑/压力位
        potential_levels = price_range[peaks]  # 获取潜在支撑/压力位的价格

        # 计算VWAP（成交量加权平均价）
        vwap_15 = np.average(closes_15, weights=volumes_15)  # 计算VWAP

        # 计算移动平均
        ma20_15 = np.mean(closes_15[-20:]) if len(closes_15) >= 20 else np.mean(closes_15)  # 计算20周期移动平均
        ma50_15 = np.mean(closes_15[-50:]) if len(closes_15) >= 50 else np.mean(closes_15)  # 计算50周期移动平均

        # 使用5分钟数据验证和细化
        closes_5 = data_5min['close'].values  # 获取5分钟数据的收盘价
        volumes_5 = data_5min['volume'].values  # 获取5分钟数据的成交量

        # 计算支撑位和压力位
        support_levels = {}  # 初始化支撑位字典
        resistance_levels = {}  # 初始化压力位字典

        # 添加基于价格分布的支撑位
        for level in potential_levels:
            if level < current_price:  # 如果潜在支撑位低于当前价格
                # 验证5分钟数据中的支撑情况
                near_level = (data_5min['low'] >= level * 0.998) & (data_5min['low'] <= level * 1.002)  # 寻找5分钟数据中接近潜在支撑位的价格
                if near_level.any():  # 如果存在接近潜在支撑位的价格
                    support_levels[f'支撑位_价格密集'] = level  # 将其添加到支撑位字典

        # 添加基于价格分布的压力位
        for level in potential_levels:
            if level > current_price:  # 如果潜在压力位高于当前价格
                # 验证5分钟数据中的压力情况
                near_level = (data_5min['high'] <= level * 1.002) & (data_5min['high'] >= level * 0.998)  # 寻找5分钟数据中接近潜在压力位的价格
                if near_level.any():  # 如果存在接近潜在压力位的价格
                    resistance_levels[f'压力位_价格密集'] = level  # 将其添加到压力位字典

        # 添加VWAP支撑压力
        if vwap_15 < current_price:  # 如果VWAP低于当前价格，则作为支撑位
            support_levels['支撑位_VWAP'] = vwap_15
        else:  # 否则作为压力位
            resistance_levels['压力位_VWAP'] = vwap_15

        # 添加移动平均支撑压力
        if ma20_15 < current_price:  # 如果20周期移动平均低于当前价格，则作为支撑位
            support_levels['支撑位_MA20'] = ma20_15
        else:  # 否则作为压力位
            resistance_levels['压力位_MA20'] = ma20_15

        if ma50_15 < current_price:  # 如果50周期移动平均低于当前价格，则作为支撑位
            support_levels['支撑位_MA50'] = ma50_15
        else:  # 否则作为压力位
            resistance_levels['压力位_MA50'] = ma50_15

        return {
            '支撑位': support_levels,  # 返回支撑位字典
            '压力位': resistance_levels,  # 返回压力位字典
            '趋势': 'up' if ma20_15 > ma50_15 else 'down'  # 根据移动平均判断趋势
        }

    def calculate_efficacy(self, levels, test_15min, test_5min):
        """
        计算支撑压力位有效性

        参数：
            levels (dict): 包含支撑位和压力位的字典。
            test_15min (pd.DataFrame): 15分钟测试数据。
            test_5min (pd.DataFrame): 5分钟测试数据。

        返回：
            dict: 包含支撑位和压力位有效性信息的字典。
        """
        efficacy = {'支撑位': {}, '压力位': {}}  # 初始化有效性字典

        if len(test_15min) < 3 or len(test_5min) < 3:  # 如果测试数据不足，返回空结果
            return efficacy

        def calculate_atr(df):
            """计算平均真实波动范围 (ATR)"""
            high = df['high']  # 获取最高价
            low = df['low']  # 获取最低价
            close = df['close']  # 获取收盘价
            tr1 = high - low  # 计算当日真实波动范围
            tr2 = abs(high - close.shift(1))  # 计算当日最高价与昨日收盘价之差的绝对值
            tr3 = abs(low - close.shift(1))  # 计算当日最低价与昨日收盘价之差的绝对值
            tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)  # 取三者最大值作为当日真实波动范围
            return tr.mean()  # 返回平均真实波动范围

        # 计算动态阈值
        atr_15 = calculate_atr(test_15min)  # 计算15分钟数据的ATR
        atr_5 = calculate_atr(test_5min)  # 计算5分钟数据的ATR
        threshold_15 = min(max(atr_15 / test_15min['close'].mean(), 0.001), 0.005)  # 计算15分钟数据的阈值
        threshold_5 = min(max(atr_5 / test_5min['close'].mean(), 0.0005), 0.003)  # 计算5分钟数据的阈值

        for level_type in ['支撑位', '压力位']:  # 遍历支撑位和压力位
            level_dict = levels[level_type]  # 获取支撑位或压力位字典
            for key, level_price in level_dict.items():  # 遍历支撑位或压力位
                is_valid = False  # 初始化有效性标志

                if level_type == '支撑位':  # 如果是支撑位
                    # 15分钟验证
                    touched_15 = (test_15min['low'] <= level_price * (1 + threshold_15)).any()  # 检查15分钟数据中是否有价格触及支撑位
                    if touched_15:  # 如果触及支撑位
                        idx_15 = test_15min[test_15min['low'] <= level_price * (1 + threshold_15)].index[0]  # 获取触及支撑位的索引
                        post_touch_15 = test_15min.loc[idx_15:]  # 获取触及支撑位之后的数据

                        # 5分钟精确验证
                        touch_time = test_15min.loc[idx_15, 'date']  # 获取触及支撑位的时间
                        post_touch_5 = test_5min[test_5min['date'] >= touch_time]  # 获取触及支撑位之后的数据

                        if len(post_touch_5) > 0:  # 如果有触及支撑位之后的数据
                            bounce_5 = (post_touch_5['high'].max() - post_touch_5['low'].min()) / post_touch_5['low'].min()  # 计算反弹幅度
                            volume_surge = (post_touch_5['volume'] > post_touch_5['volume'].mean() * 1.2).any()  # 检查是否有成交量激增
                            is_valid = bounce_5 > threshold_5 * 2 and volume_surge  # 判断支撑位是否有效

                else:  # 如果是压力位
                    # 15分钟验证
                    touched_15 = (test_15min['high'] >= level_price * (1 - threshold_15)).any()  # 检查15分钟数据中是否有价格触及压力位
                    if touched_15:  # 如果触及压力位
                        idx_15 = test_15min[test_15min['high'] >= level_price * (1 - threshold_15)].index[0]  # 获取触及压力位的索引
                        post_touch_15 = test_15min.loc[idx_15:]  # 获取触及压力位之后的数据

                        # 5分钟精确验证
                        touch_time = test_15min.loc[idx_15, 'date']  # 获取触及压力位的时间
                        post_touch_5 = test_5min[test_5min['date'] >= touch_time]  # 获取触及压力位之后的数据

                        if len(post_touch_5) > 0:  # 如果有触及压力位之后的数据
                            drop_5 = (post_touch_5['high'].max() - post_touch_5['low'].min()) / post_touch_5['high'].max()  # 计算下跌幅度
                            volume_surge = (post_touch_5['volume'] > post_touch_5['volume'].mean() * 1.2).any()  # 检查是否有成交量激增
                            is_valid = drop_5 > threshold_5 * 2 and volume_surge  # 判断压力位是否有效

                efficacy[level_type][key] = is_valid  # 将支撑位或压力位的有效性信息添加到有效性字典

        return efficacy  # 返回有效性字典

    def run_backtest(self):
        """运行回测"""
        self.results = []  # 初始化结果列表

        for window_size in self.window_sizes:  # 遍历每个窗口大小
            print(f"\n正在测试 {window_size} 小时窗口...")
            window_results = []  # 初始化窗口结果列表

            # 将15分钟数据按小时分组
            hours = (self.df_15min['date'].max() - self.df_15min['date'].min()).total_seconds() / 3600  # 计算数据的时间跨度（以小时为单位）
            test_periods = int(hours - window_size)  # 计算测试周期数

            for i in tqdm(range(test_periods)):  # 遍历每个测试周期
                end_time = self.df_15min['date'].max() - pd.Timedelta(hours=i)  # 计算训练数据的结束时间
                start_time = end_time - pd.Timedelta(hours=window_size)  # 计算训练数据的起始时间
                test_start = end_time  # 计算测试数据的起始时间
                test_end = end_time + pd.Timedelta(hours=1)  # 计算测试数据的结束时间

                # 获取训练数据
                train_15min = self.df_15min[(self.df_15min['date'] >= start_time) &
                                          (self.df_15min['date'] < end_time)]  # 获取15分钟训练数据
                train_5min = self.df_5min[(self.df_5min['date'] >= start_time) &
                                          (self.df_5min['date'] < end_time)]  # 获取5分钟训练数据

                # 获取测试数据
                test_15min = self.df_15min[(self.df_15min['date'] >= test_start) &
                                         (self.df_15min['date'] < test_end)]  # 获取15分钟测试数据
                test_5min = self.df_5min[(self.df_5min['date'] >= test_start) &
                                       (self.df_5min['date'] < test_end)]  # 获取5分钟测试数据

                if len(train_15min) == 0 or len(test_15min) == 0:  # 如果训练数据或测试数据为空，跳过本次循环
                    continue

                # 计算支撑压力位
                levels = self.calculate_levels(train_15min, train_5min)  # 计算支撑压力位

                # 验证支撑压力位有效性
                efficacy = self.calculate_efficacy(levels, test_15min, test_5min)  # 计算支撑压力位有效性

                window_results.append({  # 将结果添加到窗口结果列表
                    '训练开始': start_time,  # 训练数据的起始时间
                    '训练结束': end_time,  # 训练数据的结束时间
                    '测试时间': test_start,  # 测试数据的起始时间
                    '支撑压力位': levels,  # 支撑压力位
                    '验证结果': efficacy  # 支撑压力位有效性
                })

            self.results.append({  # 将窗口结果添加到结果列表
                '窗口大小': window_size,  # 窗口大小
                '结果': window_results  # 窗口结果列表
            })

            # 分析该窗口大小的整体表现
            self.analyze_window_performance(window_results)  # 分析窗口表现

    def analyze_window_performance(self, results):
        """
        分析窗口表现

        参数：
            results (list): 窗口结果列表。
        """
        total_support = 0  # 初始化支撑位总数
        valid_support = 0  # 初始化有效支撑位数
        total_resistance = 0  # 初始化压力位总数
        valid_resistance = 0  # 初始化有效压力位数

        for result in results:  # 遍历每个结果
            efficacy = result['验证结果']  # 获取支撑压力位有效性信息

            for key, valid in efficacy['支撑位'].items():  # 遍历支撑位
                total_support += 1  # 累加支撑位总数
                if valid:  # 如果支撑位有效
                    valid_support += 1  # 累加有效支撑位数

            for key, valid in efficacy['压力位'].items():  # 遍历压力位
                total_resistance += 1  # 累加压力位总数
                if valid:  # 如果压力位有效
                    valid_resistance += 1  # 累加有效压力位数

        print("\n窗口表现分析:")
        print(f"总体样本数: {len(results)}")  # 打印样本总数
        print(f"支撑位准确率: {valid_support/total_support*100:.2f}% (样本数: {total_support})" if total_support > 0 else "支撑位样本不足")  # 打印支撑位准确率
        print(f"压力位准确率: {valid_resistance/total_resistance*100:.2f}% (样本数: {total_resistance})" if total_resistance > 0 else "压力位样本不足")  # 打印压力位准确率

    def plot_results(self):
        """绘制支撑压力位和价格走势"""
        import matplotlib.pyplot as plt
        from matplotlib.gridspec import GridSpec

        # 获取最近24小时的数据用于绘图
        last_24h = self.df_15min[self.df_15min['date'] >= (self.df_15min['date'].max() - pd.Timedelta(hours=24))]

        # 创建图表
        fig = plt.figure(figsize=(15, 10))
        gs = GridSpec(2, 1, height_ratios=[3, 1])

        # 价格走势和支撑压力位
        ax1 = fig.add_subplot(gs[0])
        ax2 = fig.add_subplot(gs[1])

        # 绘制价格走势
        ax1.plot(last_24h['date'], last_24h['close'], label='价格', color='black', alpha=0.7)

        # 获取最新的支撑压力位
        latest_levels = self.calculate_levels(
            last_24h,
            self.df_5min[self.df_5min['date'] >= (self.df_5min['date'].max() - pd.Timedelta(hours=24))]
        )

        # 绘制支撑位
        for key, level in latest_levels['支撑位'].items():
            ax1.axhline(y=level, color='g', linestyle='--', alpha=0.5, label=f'{key}: {level:.2f}')

        # 绘制压力位
        for key, level in latest_levels['压力位'].items():
            ax1.axhline(y=level, color='r', linestyle='--', alpha=0.5, label=f'{key}: {level:.2f}')

        # 绘制成交量
        ax2.bar(last_24h['date'], last_24h['volume'], color='blue', alpha=0.3)

        # 设置图表格式
        ax1.set_title('价格走势与支撑压力位 (最近24小时)')
        ax1.set_xlabel('时间')
        ax1.set_ylabel('价格')
        ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        ax1.grid(True, alpha=0.3)

        ax2.set_title('成交量')
        ax2.set_xlabel('时间')
        ax2.set_ylabel('成交量')
        ax2.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def print_latest_result(self):
        """打印最新的支撑压力位预测结果"""
        # 获取最近数据
        latest_24h = self.df_15min[self.df_15min['date'] >= (self.df_15min['date'].max() - pd.Timedelta(hours=24))]
        latest_24h_5min = self.df_5min[self.df_5min['date'] >= (self.df_5min['date'].max() - pd.Timedelta(hours=24))]

        # 计算最新的支撑压力位
        latest_levels = self.calculate_levels(latest_24h, latest_24h_5min)
        current_price = latest_24h['close'].iloc[-1]

        print("\n=== 最新预测结果 ===")
        print(f"当前价格: {current_price:.2f}")
        print(f"市场趋势: {latest_levels['趋势']}")
        print("\n支撑位:")
        for key, level in latest_levels['支撑位'].items():
            distance = (current_price - level) / current_price * 100
            print(f"{key}: {level:.2f} (距当前价格: {distance:.2f}%)")

        print("\n压力位:")
        for key, level in latest_levels['压力位'].items():
            distance = (level - current_price) / current_price * 100
            print(f"{key}: {level:.2f} (距当前价格: {distance:.2f}%)")

        # 打印准确率统计
        print("\n=== 准确率统计 ===")
        for result in self.results:
            window_size = result['窗口大小']
            window_results = result['结果']

            total_support = sum(len(r['验证结果']['支撑位']) for r in window_results)
            valid_support = sum(sum(v for v in r['验证结果']['支撑位'].values()) for r in window_results)

            total_resistance = sum(len(r['验证结果']['压力位']) for r in window_results)
            valid_resistance = sum(sum(v for v in r['验证结果']['压力位'].values()) for r in window_results)

            print(f"\n{window_size}小时窗口:")
            print(f"支撑位准确率: {valid_support/total_support*100:.2f}% (样本数: {total_support})")
            print(f"压力位准确率: {valid_resistance/total_resistance*100:.2f}% (样本数: {total_resistance})")




tester = SupportPressureTester(
    min15_path='2228_data_15min.csv',
    min5_path='2228_data_5min.csv',
    window_sizes=[24, 48, 72]  # 使用小时作为窗口单位
)
tester.run_backtest()
tester.plot_results()
tester.print_latest_result()
```

```python
from IPython import get_ipython  # 导入IPython库，用于交互式环境
from IPython.display import display  # 导入display函数，用于显示内容

# %%
!wget https://raw.githubusercontent.com/StellarCN/scp_zh/refs/heads/master/fonts/SimHei.ttf  # 下载SimHei字体文件

import numpy as np  # 导入numpy库，用于数值计算
import pandas as pd  # 导入pandas库，用于数据处理
from datetime import date  # 导入date模块，用于处理日期
from scipy import stats, signal  # 导入scipy库的stats和signal模块，用于统计分析和信号处理
from tqdm import tqdm  # 导入tqdm库，用于显示进度条

import matplotlib.pyplot as plt  # 导入matplotlib库，用于绘图
import matplotlib.font_manager as fm  # 导入font_manager模块，用于管理字体
from datetime import date, datetime  # 导入date和datetime模块，用于处理日期和时间

# 设置中文字体
font_path = 'SimHei.ttf'  # 设置字体路径，请确保SimHei.ttf字体文件在您的环境中可用
fontprop = fm.FontProperties(fname=font_path, family='SimHei')  # 创建字体属性对象

# 刷新字体缓存
fm.fontManager.addfont(font_path)  # 将字体添加到字体管理器

# 设置SimHei为默认字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置全局字体为SimHei
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

class SupportPressureTester:
    def __init__(self, min15_path, min5_path, window_sizes=[24, 48, 72]):
        """初始化测试器"""
        # 读取分时数据
        self.df_15min = pd.read_csv(min15_path)  # 读取15分钟数据
        self.df_5min = pd.read_csv(min5_path)  # 读取5分钟数据

        # 确保日期格式正确
        self.df_15min['date'] = pd.to_datetime(self.df_15min['date'])  # 将日期列转换为datetime格式
        self.df_5min['date'] = pd.to_datetime(self.df_5min['date'])  # 将日期列转换为datetime格式

        # 按日期排序
        self.df_15min = self.df_15min.sort_values('date').reset_index(drop=True)  # 按日期排序并重置索引
        self.df_5min = self.df_5min.sort_values('date').reset_index(drop=True)  # 按日期排序并重置索引

        self.window_sizes = window_sizes  # 设置窗口大小
        self.results = []  # 初始化结果列表

        print(f"数据范围: {self.df_15min['date'].min().strftime('%Y-%m-%d %H:%M')} 至 {self.df_15min['date'].max().strftime('%Y-%m-%d %H:%M')}") # 打印数据范围

    def calculate_psychological_levels(self, current_price, range_percent=0.05):
        """计算心理关口位"""
        levels = {}  # 初始化关口位字典

        # 确定价格范围
        price_range_low = current_price * (1 - range_percent)  # 计算价格范围下限
        price_range_high = current_price * (1 + range_percent)  # 计算价格范围上限

        # 生成可能的心理关口
        thousands = [i * 1000 for i in range(int(price_range_low/1000), int(price_range_high/1000) + 1)]  # 生成整千位关口
        hundreds = [i * 100 for i in range(int(price_range_low/100), int(price_range_high/100) + 1)]  # 生成整百位关口
        fifties = [i * 50 for i in range(int(price_range_low/50), int(price_range_high/50) + 1)]  # 生成半位关口
        special = [5000, 5500, 6000, 6500, 7000]  # 特殊关口

        all_levels = {
            '整千位': thousands,
            '整百位': hundreds,
            '半位': fifties,
            '特殊关口': special
        }  # 将所有关口类型和对应的关口值存储在字典中

        for level_type, numbers in all_levels.items():  # 遍历所有关口类型
            for number in numbers:  # 遍历该类型下的所有关口值
                if price_range_low <= number <= price_range_high:  # 如果关口值在价格范围内
                    if number < current_price:  # 如果关口值小于当前价格，则为支撑位
                        levels[f'支撑位_心理_{level_type}_{number}'] = number
                    elif number > current_price:  # 如果关口值大于当前价格，则为压力位
                        levels[f'压力位_心理_{level_type}_{number}'] = number

        return levels  # 返回关口位字典

    def calculate_fibonacci_projections(self, data, is_breakout_high=True):
        """计算斐波那契投射位"""
        high = data['high'].max()  # 获取最高价
        low = data['low'].min()  # 获取最低价
        diff = high - low  # 计算价差

        if is_breakout_high:  # 如果突破高点
            return {
                '压力位_Fib_1.272': high + diff * 0.272,  # 计算1.272倍斐波那契压力位
                '压力位_Fib_1.414': high + diff * 0.414,  # 计算1.414倍斐波那契压力位
                '压力位_Fib_1.618': high + diff * 0.618   # 计算1.618倍斐波那契压力位
            }
        else:  # 如果突破低点
            return {
                '支撑位_Fib_0.786': low - diff * 0.214,  # 计算0.786倍斐波那契支撑位
                '支撑位_Fib_0.618': low - diff * 0.382,  # 计算0.618倍斐波那契支撑位
                '支撑位_Fib_0.5': low - diff * 0.5      # 计算0.5倍斐波那契支撑位
            }

    def calculate_volatility_based_levels(self, data, is_breakout_high=True):
        """基于波动率计算潜在位置"""
        high = data['high']  # 获取最高价序列
        low = data['low']  # 获取最低价序列
        close = data['close'].shift(1)  # 获取前一周期收盘价序列
        tr1 = high - low  # 计算当日最高价与最低价之差
        tr2 = abs(high - close)  # 计算当日最高价与前一周期收盘价之差的绝对值
        tr3 = abs(low - close)  # 计算当日最低价与前一周期收盘价之差的绝对值
        tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)  # 计算真实波动幅度TR
        atr = tr.rolling(window=14).mean().iloc[-1]  # 计算平均真实波动幅度ATR

        current_price = data['close'].iloc[-1]  # 获取当前价格

        if is_breakout_high:  # 如果突破高点
            return {
                '压力位_ATR_1': current_price + atr,  # 计算基于ATR的压力位1
                '压力位_ATR_2': current_price + atr * 2,  # 计算基于ATR的压力位2
                '压力位_ATR_3': current_price + atr * 3   # 计算基于ATR的压力位3
            }
        else:  # 如果突破低点
            return {
                '支撑位_ATR_1': current_price - atr,  # 计算基于ATR的支撑位1
                '支撑位_ATR_2': current_price - atr * 2,  # 计算基于ATR的支撑位2
                '支撑位_ATR_3': current_price - atr * 3   # 计算基于ATR的支撑位3
            }

    def calculate_traditional_levels(self, data_15min, data_5min):
        """计算传统支撑压力位"""
        if len(data_15min) == 0 or len(data_5min) == 0:  # 如果数据为空，则返回空字典
            return {'支撑位': {}, '压力位': {}, '趋势': 'unknown'}

        # 使用15分钟数据计算
        highs = data_15min['high'].values  # 获取最高价序列
        lows = data_15min['low'].values  # 获取最低价序列
        closes = data_15min['close'].values  # 获取收盘价序列
        volumes = data_15min['volume'].values  # 获取成交量序列
        current_price = closes[-1]  # 获取当前价格

        # 计算VWAP
        vwap = np.average(closes, weights=volumes)  # 计算成交量加权平均价

        # 计算移动平均
        ma20 = np.mean(closes[-20:]) if len(closes) >= 20 else np.mean(closes)  # 计算20周期移动平均线
        ma50 = np.mean(closes[-50:]) if len(closes) >= 50 else np.mean(closes)  # 计算50周期移动平均线

        levels = {'支撑位': {}, '压力位': {}, '趋势': 'up' if ma20 > ma50 else 'down'}  # 初始化关口位字典，并根据移动平均线判断趋势

        # 分配VWAP和MA
        if vwap < current_price:  # 如果VWAP小于当前价格，则为支撑位
            levels['支撑位']['支撑位_VWAP'] = vwap
        else:  # 否则为压力位
            levels['压力位']['压力位_VWAP'] = vwap

        if ma20 < current_price:  # 如果MA20小于当前价格，则为支撑位
            levels['支撑位']['支撑位_MA20'] = ma20
        else:  # 否则为压力位
            levels['压力位']['压力位_MA20'] = ma20

        if ma50 < current_price:  # 如果MA50小于当前价格，则为支撑位
            levels['支撑位']['支撑位_MA50'] = ma50
        else:  # 否则为压力位
            levels['压力位']['压力位_MA50'] = ma50

        return levels  # 返回关口位字典

    def calculate_levels(self, data_15min, data_5min):
        """计算所有支撑压力位"""
        # 首先计算传统支撑压力位
        levels = self.calculate_traditional_levels(data_15min, data_5min)  # 计算传统关口位
        current_price = data_15min['close'].iloc[-1]  # 获取当前价格

        # 检查是否突破新高或新低
        is_breakout_high = current_price >= data_15min['high'].max()  # 判断是否突破新高
        is_breakout_low = current_price <= data_15min['low'].min()  # 判断是否突破新低

        # 添加心理关口
        psychological_levels = self.calculate_psychological_levels(current_price)  # 计算心理关口位
        for key, value in psychological_levels.items():  # 遍历心理关口位
            if '支撑位' in key:  # 如果是支撑位
                levels['支撑位'][key] = value
            else:  # 如果是压力位
                levels['压力位'][key] = value

        if is_breakout_high:  # 如果突破新高
            # 添加突破高点时的额外压力位
            levels['压力位'].update(self.calculate_fibonacci_projections(data_15min, True))  # 添加斐波那契投射压力位
            levels['压力位'].update(self.calculate_volatility_based_levels(data_15min, True))  # 添加基于波动率的压力位

        if is_breakout_low:  # 如果突破新低
            # 添加突破低点时的额外支撑位
            levels['支撑位'].update(self.calculate_fibonacci_projections(data_15min, False))  # 添加斐波那契投射支撑位
            levels['支撑位'].update(self.calculate_volatility_based_levels(data_15min, False))  # 添加基于波动率的支撑位

        # 过滤太接近的位置
        def filter_levels(levels_dict):  # 定义过滤函数
            sorted_levels = dict(sorted(levels_dict.items(), key=lambda x: x[1]))  # 按关口值排序
            filtered_levels = {}  # 初始化过滤后的关口位字典
            last_level = None  # 初始化上一个关口位
            min_distance = current_price * 0.002  # 设置最小距离阈值

            for key, value in sorted_levels.items():  # 遍历排序后的关口位
                if last_level is None or abs(value - last_level) >= min_distance:  # 如果当前关口位与上一个关口位的距离大于阈值
                    filtered_levels[key] = value  # 将当前关口位添加到过滤后的字典中
                    last_level = value  # 更新上一个关口位

            return filtered_levels  # 返回过滤后的关口位字典

        levels['支撑位'] = filter_levels(levels['支撑位'])  # 过滤支撑位
        levels['压力位'] = filter_levels(levels['压力位'])  # 过滤压力位

        return levels  # 返回所有关口位

    def filter_levels(levels_dict):
        """过滤太接近的支撑/压力位"""
        sorted_levels = dict(sorted(levels_dict.items(), key=lambda x: x[1]))  # 对支撑/压力位按值排序
        filtered_levels = {}  # 初始化过滤后的字典
        last_level = None  # 初始化上一个支撑/压力位
        # 增加最小距离阈值
        min_distance = current_price * 0.005  # 设置最小距离阈值，防止支撑/压力位过于接近

        for key, value in sorted_levels.items():  # 遍历排序后的支撑/压力位
            if last_level is None or abs(value - last_level) >= min_distance:  # 如果当前支撑/压力位与上一个支撑/压力位的距离大于阈值
                filtered_levels[key] = value  # 将当前支撑/压力位添加到过滤后的字典中
                last_level = value  # 更新上一个支撑/压力位

        return filtered_levels  # 返回过滤后的字典

    def calculate_efficacy(self, levels, test_15min, test_5min):
        """计算支撑压力位有效性"""
        efficacy = {'支撑位': {}, '压力位': {}}  # 初始化有效性字典

        if len(test_15min) < 3 or len(test_5min) < 3:  # 如果测试数据不足，则返回空字典
            return efficacy

        def calculate_atr(df):  # 定义计算ATR的函数
            high = df['high']  # 获取最高价序列
            low = df['low']  # 获取最低价序列
            close = df['close']  # 获取收盘价序列
            tr1 = high - low  # 计算当日最高价与最低价之差
            tr2 = abs(high - close.shift(1))  # 计算当日最高价与前一周期收盘价之差的绝对值
            tr3 = abs(low - close.shift(1))  # 计算当日最低价与前一周期收盘价之差的绝对值
            tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)  # 计算真实波动幅度TR
            return tr.mean()  # 返回平均真实波动幅度ATR

        atr_15 = calculate_atr(test_15min)  # 计算15分钟数据的ATR
        atr_5 = calculate_atr(test_5min)  # 计算5分钟数据的ATR
        threshold_15 = min(max(atr_15 / test_15min['close'].mean(), 0.001), 0.005)  # 计算15分钟数据的阈值
        threshold_5 = min(max(atr_5 / test_5min['close'].mean(), 0.0005), 0.003)  # 计算5分钟数据的阈值

        for level_type in ['支撑位', '压力位']:  # 遍历支撑位和压力位
            level_dict = levels[level_type]  # 获取当前类型的关口位字典
            for key, level_price in level_dict.items():  # 遍历关口位
                is_valid = False  # 初始化有效性标志

                if level_type == '支撑位':  # 如果是支撑位
                    touched_15 = (test_15min['low'] <= level_price * (1 + threshold_15)).any()  # 判断15分钟数据是否触及支撑位
                    if touched_15:  # 如果触及支撑位
                        idx_15 = test_15min[test_15min['low'] <= level_price * (1 + threshold_15)].index[0]  # 获取触及支撑位的索引
                        touch_time = test_15min.loc[idx_15, 'date']  # 获取触及支撑位的时间
                        post_touch_5 = test_5min[test_5min['date'] >= touch_time]  # 获取触及支撑位之后的数据
                        if len(post_touch_5) > 0:  # 如果有触及支撑位之后的数据

                            bounce_5 = (post_touch_5['high'].max() - post_touch_5['low'].min()) / post_touch_5['low'].min()  # 计算反弹幅度
                            volume_surge = (post_touch_5['volume'] > post_touch_5['volume'].mean() * 1.2).any()  # 判断是否有成交量异动
                            is_valid = bounce_5 > threshold_5 * 2 and volume_surge  # 判断支撑位是否有效

                else:  # 如果是压力位
                    touched_15 = (test_15min['high'] >= level_price * (1 - threshold_15)).any()  # 判断15分钟数据是否触及压力位
                    if touched_15:  # 如果触及压力位
                        idx_15 = test_15min[test_15min['high'] >= level_price * (1 - threshold_15)].index[0]  # 获取触及压力位的索引
                        touch_time = test_15min.loc[idx_15, 'date']  # 获取触及压力位的时间
                        post_touch_5 = test_5min[test_5min['date'] >= touch_time]  # 获取触及压力位之后的数据
                        if len(post_touch_5) > 0:  # 如果有触及压力位之后的数据
                            drop_5 = (post_touch_5['high'].max() - post_touch_5['low'].min()) / post_touch_5['high'].max()  # 计算下跌幅度
                            volume_surge = (post_touch_5['volume'] > post_touch_5['volume'].mean() * 1.2).any()  # 判断是否有成交量异动
                            is_valid = drop_5 > threshold_5 * 2 and volume_surge  # 判断压力位是否有效

                efficacy[level_type][key] = is_valid  # 将有效性标志存储到字典中

        return efficacy  # 返回有效性字典

    def run_backtest(self):
        """运行回测"""
        self.results = []  # 初始化结果列表

        for window_size in self.window_sizes:  # 遍历窗口大小
            print(f"\n正在测试 {window_size} 小时窗口...")
            window_results = []  # 初始化窗口结果列表

            hours = (self.df_15min['date'].max() - self.df_15min['date'].min()).total_seconds() / 3600  # 计算总小时数
            test_periods = int(hours - window_size)  # 计算测试周期数

            for i in tqdm(range(test_periods)):  # 遍历测试周期，使用tqdm显示进度条
                end_time = self.df_15min['date'].max() - pd.Timedelta(hours=i)  # 计算训练结束时间
                start_time = end_time - pd.Timedelta(hours=window_size)  # 计算训练开始时间
                test_start = end_time  # 计算测试开始时间
                test_end = end_time + pd.Timedelta(hours=1)  # 计算测试结束时间

                # 获取训练数据
                train_15min = self.df_15min[(self.df_15min['date'] >= start_time) &
                                          (self.df_15min['date'] < end_time)]  # 获取15分钟训练数据
                train_5min = self.df_5min[(self.df_5min['date'] >= start_time) &
                                        (self.df_5min['date'] < end_time)]  # 获取5分钟训练数据

                # 获取测试数据
                test_15min = self.df_15min[(self.df_15min['date'] >= test_start) &
                                        (self.df_15min['date'] < test_end)]  # 获取15分钟测试数据
                test_5min = self.df_5min[(self.df_5min['date'] >= test_start) &
                                      (self.df_5min['date'] < test_end)]  # 获取5分钟测试数据

                if len(train_15min) == 0 or len(test_15min) == 0:  # 如果训练数据或测试数据为空，则跳过
                    continue

                levels = self.calculate_levels(train_15min, train_5min)  # 计算支撑压力位
                efficacy = self.calculate_efficacy(levels, test_15min, test_5min)  # 计算支撑压力位有效性

                window_results.append({  # 将结果添加到窗口结果列表中
                    '训练开始': start_time,
                    '训练结束': end_time,
                    '测试时间': test_start,
                    '支撑压力位': levels,
                    '验证结果': efficacy
                })

            self.results.append({  # 将窗口结果添加到总结果列表中
                '窗口大小': window_size,
                '结果': window_results
            })

            self.analyze_window_performance(window_results)  # 分析窗口表现

    def analyze_window_performance(self, results):
        """分析窗口表现"""
        total_support = 0  # 初始化支撑位总数
        valid_support = 0  # 初始化有效支撑位数
        total_resistance = 0  # 初始化压力位总数
        valid_resistance = 0  # 初始化有效压力位数

        for result in results:  # 遍历结果
            efficacy = result['验证结果']  # 获取有效性字典

            for key, valid in efficacy['支撑位'].items():  # 遍历支撑位
                total_support += 1  # 累加支撑位总数
                if valid:  # 如果支撑位有效
                    valid_support += 1  # 累加有效支撑位数

            for key, valid in efficacy['压力位'].items():  # 遍历压力位
                total_resistance += 1  # 累加压力位总数
                if valid:  # 如果压力位有效
                    valid_resistance += 1  # 累加有效压力位数

        print("\n窗口表现分析:")
        print(f"总体样本数: {len(results)}")  # 打印样本总数
        if total_support > 0:  # 如果有支撑位
            print(f"支撑位准确率: {valid_support/total_support*100:.2f}% (样本数: {total_support})")  # 打印支撑位准确率
        if total_resistance > 0:  # 如果有压力位
            print(f"压力位准确率: {valid_resistance/total_resistance*100:.2f}% (样本数: {total_resistance})")  # 打印压力位准确率

    def print_latest_result(self):
        """打印最新的支撑压力位预测结果"""
        latest_24h = self.df_15min[self.df_15min['date'] >= (self.df_15min['date'].max() - pd.Timedelta(hours=24))]  # 获取最近24小时的15分钟数据
        latest_24h_5min = self.df_5min[self.df_5min['date'] >= (self.df_5min['date'].max() - pd.Timedelta(hours=24))]  # 获取最近24小时的5分钟数据

        latest_levels = self.calculate_levels(latest_24h, latest_24h_5min)  # 计算最新的支撑压力位
        current_price= latest_24h['close'].iloc[-1]  # 获取当前价格

        print("\n=== 最新预测结果 ===")
        print(f"当前价格: {current_price:.2f}")  # 打印当前价格
        print(f"市场趋势: {latest_levels['趋势']}")  # 打印市场趋势

    def print_levels(levels, level_type):  # 定义打印支撑/压力位的函数
        if levels:  # 如果有支撑/压力位
            print(f"\n{level_type}:")  # 打印支撑/压力位类型
            sorted_levels = sorted(levels.items(),
                                  key=lambda x: abs(x[1] - current_price))  # 按距离当前价格排序

            for key, level in sorted_levels:  # 遍历支撑/压力位
                if level_type == "支撑位":  # 如果是支撑位
                    distance = (current_price - level) / current_price * 100  # 计算距离当前价格的百分比
                else:  # 如果是压力位
                    distance = (level - current_price) / current_price * 100  # 计算距离当前价格的百分比

                strength = "强" if "心理" in key or "VWAP" in key else "中"  # 根据类型判断强度
                if distance < 0.5:  # 如果距离当前价格很近
                    strength += "+"  # 增加强度

                print(f"{key}: {level:.2f} (距当前价格: {distance:.2f}%, 强度: {strength})")  # 打印支撑/压力位信息

        print_levels(latest_levels['支撑位'], "支撑位")  # 打印支撑位
        print_levels(latest_levels['压力位'], "压力位")  # 打印压力位

    def plot_results(self):
        """绘制支撑压力位和价格走势"""
        import matplotlib.pyplot as plt  # 导入matplotlib库
        from matplotlib.gridspec import GridSpec  # 导入GridSpec用于创建子图
        import matplotlib.dates as mdates  # 导入mdates用于处理日期格式
        from matplotlib.patches import Rectangle  # 导入Rectangle用于绘制矩形

        # 获取最近24小时的数据用于绘图
        last_24h = self.df_15min[self.df_15min['date'] >= (self.df_15min['date'].max() - pd.Timedelta(hours=24))]  # 获取最近24小时的15分钟数据

        # 创建图表
        #plt.style.use('classic')
        fig = plt.figure(figsize=(15, 10))  # 创建图表对象，设置尺寸
        gs = GridSpec(2, 1, height_ratios=[3, 1])  # 创建GridSpec对象，用于划分图表区域

        # 设置背景色
        fig.patch.set_facecolor('#F0F3F5')  # 设置图表背景色

        # 价格走势和支撑压力位
        ax1 = fig.add_subplot(gs[0])  # 创建第一个子图，用于绘制价格走势和支撑压力位
        ax2 = fig.add_subplot(gs[1])  # 创建第二个子图，用于绘制成交量

        # 设置子图背景色
        ax1.set_facecolor('#FFFFFF')  # 设置第一个子图背景色
        ax2.set_facecolor('#FFFFFF')  # 设置第二个子图背景色

        # 添加网格
        ax1.grid(True, linestyle='--', alpha=0.3)  # 为第一个子图添加网格
        ax2.grid(True, linestyle='--', alpha=0.3)  # 为第二个子图添加网格

        # 绘制K线图
        ax1.plot(last_24h['date'], last_24h['close'],
                color='#2C3E50', linewidth=1.5, label='收盘价',
                zorder=5)  # 绘制收盘价曲线

        # 绘制价格区间
        ax1.fill_between(last_24h['date'], last_24h['high'], last_24h['low'],
                        color='#BDC3C7', alpha=0.3, label='价格区间',
                        zorder=4)  # 绘制价格区间

        # 获取最新的支撑压力位
        latest_levels = self.calculate_levels(
            last_24h,
            self.df_5min[self.df_5min['date'] >= (self.df_5min['date'].max() - pd.Timedelta(hours=24))]
        )  # 计算最新的支撑压力位

        # 定义支撑位和压力位的颜色
        support_colors = ['#27AE60', '#2ECC71', '#82E0AA']  # 深到浅的绿色
        resistance_colors = ['#C0392B', '#E74C3C', '#F1948A']  # 深到浅的红色

        # 绘制支撑位
        for i, (key, level) in enumerate(latest_levels['支撑位'].items()):  # 遍历支撑位
            color = support_colors[min(i, len(support_colors)-1)]  # 选择颜色
            style = '-' if '心理' in key else '--'  # 选择线条样式
            ax1.axhline(y=level, color=color, linestyle=style, alpha=0.7,
                        label=f'{key}: {level:.2f}',
                        zorder=3)  # 绘制支撑位水平线

        # 绘制压力位
        for i, (key, level) in enumerate(latest_levels['压力位'].items()):  # 遍历压力位
            color = resistance_colors[min(i, len(resistance_colors)-1)]  # 选择颜色
            style = '-' if '心理' in key else '--'  # 选择线条样式
            ax1.axhline(y=level, color=color, linestyle=style, alpha=0.7,
                        label=f'{key}: {level:.2f}',
                        zorder=3)  # 绘制压力位水平线

        # 绘制成交量柱状图
        for i in range(len(last_24h)):  # 遍历数据
            if i > 0:  # 如果不是第一个数据
                color = '#2ECC71' if last_24h['close'].iloc[i] >= last_24h['close'].iloc[i-1] else '#E74C3C'  # 根据收盘价变化选择颜色
            else:  # 如果是第一个数据
                color = '#2ECC71'  # 默认颜色

            ax2.bar(last_24h['date'].iloc[i], last_24h['volume'].iloc[i],
                    width=pd.Timedelta(minutes=10),  # 调整柱子宽度
                    color=color, alpha=0.7)  # 绘制成交量柱状图

        # 设置时间格式
        date_formatter = mdates.DateFormatter('%H:%M')  # 创建日期格式化器
        ax1.xaxis.set_major_formatter(date_formatter)  # 设置第一个子图的x轴日期格式
        ax2.xaxis.set_major_formatter(date_formatter)  # 设置第二个子图的x轴日期格式

        # 设置标题和标签
        ax1.set_title('价格走势与支撑压力位 (最近24小时)', fontsize=14, pad=20)  # 设置第一个子图的标题
        ax1.set_xlabel('时间', fontsize=12)  # 设置第一个子图的x轴标签
        ax1.set_ylabel('价格', fontsize=12)  # 设置第一个子图的y轴标签

        ax2.set_title('成交量', fontsize=14, pad=20)  # 设置第二个子图的标题
        ax2.set_xlabel('时间', fontsize=12)  # 设置第二个子图的x轴标签
        ax2.set_ylabel('成交量', fontsize=12)  # 设置第二个子图的y轴标签

        # 优化刻度标签
        for ax in [ax1, ax2]:  # 遍历子图
            ax.tick_params(axis='both', labelsize=10)  # 设置刻度标签大小
            plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')  # 旋转x轴刻度标签

        # 优化图例位置和样式
        ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10,
                  facecolor='white', edgecolor='none', framealpha= 0.8)  # 设置图例样式

        # 添加图表说明
        current_price = last_24h['close'].iloc[-1]  # 获取当前价格
        trend = latest_levels.get('趋势', 'unknown')  # 获取市场趋势
        trend_text = '上涨' if trend == 'up' else '下跌' if trend == 'down' else '未知'  # 转换市场趋势为中文

        # 使用文本框添加说明
        textstr = f'当前价格: {current_price:.2f}\n市场趋势: {trend_text}'  # 创建说明文本
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)  # 设置文本框样式
        ax1.text(0.02, 0.98, textstr, transform=ax1.transAxes,
                fontsize=12, verticalalignment='top',
                bbox=props)  # 添加文本框到图表

        plt.tight_layout()  # 调整图表布局
        plt.show()  # 显示图表

tester = SupportPressureTester(
    min15_path='2228_data_15min.csv',  # 设置15分钟数据文件路径
    min5_path='2228_data_5min.csv',  # 设置5分钟数据文件路径
    window_sizes=[24, 48, 72]  # 使用小时作为窗口单位
)
tester.run_backtest()  # 运行回测
tester.plot_results()  # 绘制结果
tester.print_latest_result()  # 打印最新预测结果
```

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import matplotlib.pyplot as plt  # 导入 matplotlib.pyplot 库，用于绘图

def analyze_price_crossing(df, time_frame_name, price_step=1):
    """
    分析价格穿越次数并绘图。

    参数:
    df: DataFrame, 输入数据
    time_frame_name: str, 时间周期名称 (例如 "1d", "5min", "15min")
    price_step: float, 价格步长
    """

    # 1. 确定价格范围
    min_price = df[['open', 'high', 'low', 'close']].min().min()  # 获取最低价
    max_price = df[['open', 'high', 'low', 'close']].max().max()  # 获取最高价

    # 价格范围上下各扩展一点，避免边界问题
    start_price = min_price * 0.98  # 设置起始价格，向下扩展 2%
    end_price = max_price * 1.02  # 设置结束价格，向上扩展 2%

    price_range = []  # 初始化价格范围列表
    current_price = start_price  # 设置当前价格为起始价格
    while current_price <= end_price:  # 循环生成价格范围
        price_range.append(current_price)  # 将当前价格添加到价格范围列表
        current_price += price_step  # 更新当前价格

    crossing_stats = {}  # 初始化存储结果的字典

    # 2. 遍历价格
    for price_level in price_range:  # 遍历价格范围
        crossing_stats[price_level] = {'crossing_times': 0, 'volume_before_crossing': 0, 'volume_after_crossing': 0}  # 初始化当前价格水平的统计信息
        daily_data = df.groupby(df['date'].dt.date)  # 按天分组

        # 3. 遍历每天的数据
        for date, daily_df in daily_data:  # 遍历每天的数据
            daily_df = daily_df.sort_values(by='date')  # 确保日内数据按时间排序
            last_close = None  # 初始化上一个收盘价
            crossing_type = None  # 记录上一次穿越类型 (up or down)
            daily_volume_before = 0  # 初始化当日穿越前成交量
            daily_volume_after = 0  # 初始化当日穿越后成交量
            volume_before_flag = True  # 标记是否在统计穿越前的成交量

            # 4. 遍历日内每条数据
            for index, row in daily_df.iterrows():  # 遍历日内每条数据
                current_close = row['close']  # 获取当前收盘价
                current_volume = row['volume']  # 获取当前成交量

                if last_close is not None:  # 如果存在上一个收盘价
                    up_crossing = (last_close <= price_level) and (current_close > price_level)  # 判断是否向上穿越
                    down_crossing = (last_close >= price_level) and (current_close < price_level)  # 判断是否向下穿越

                    if up_crossing:  # 如果向上穿越
                        if crossing_type == 'down':  # 如果上一次是向下穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'up'  # 更新穿越类型为向上
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                    elif down_crossing:  # 如果向下穿越
                        if crossing_type == 'up':  # 如果上一次是向上穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'down'  # 更新穿越类型为向下
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                if volume_before_flag:  # 如果在统计穿越前的成交量
                    daily_volume_before += current_volume  # 累加穿越前的成交量

                last_close = current_close  # 更新上一个收盘价

    # 5. 结果可视化
    prices = list(crossing_stats.keys())  # 获取价格列表
    crossing_counts = [crossing_stats[p]['crossing_times'] for p in prices]  # 获取穿越次数列表
    volume_before = [crossing_stats[p]['volume_before_crossing'] for p in prices]  # 获取穿越前累计成交量列表
    volume_after = [crossing_stats[p]['volume_after_crossing'] for p in prices]  # 获取穿越后累计成交量列表

    plt.figure(figsize=(12, 6))  # 创建图形对象，设置尺寸
    plt.plot(prices, crossing_counts, label='穿越次数')  # 绘制穿越次数折线图
    plt.xlabel('价格水平')  # 设置 x 轴标签
    plt.ylabel('穿越次数')  # 设置 y 轴标签
    plt.title(f'{time_frame_name} - 不同价格水平的反复穿越次数')  # 设置标题
    plt.legend()  # 显示图例
    plt.grid(True)  # 显示网格
    plt.show()  # 显示图形

    plt.figure(figsize=(12, 6))  # 创建图形对象，设置尺寸
    plt.plot(prices, volume_before, label='穿越前累计成交量', color='green')  # 绘制穿越前累计成交量折线图
    plt.plot(prices, volume_after, label='穿越后累计成交量', color='red')  # 绘制穿越后累计成交量折线图
    plt.xlabel('价格水平')  # 设置 x 轴标签
    plt.ylabel('累计成交量')  # 设置 y 轴标签
    plt.title(f'{time_frame_name} - 不同价格水平穿越前后累计成交量')  # 设置标题
    plt.legend()  # 显示图例
    plt.grid(True)  # 显示网格
    plt.show()  # 显示图形

    return crossing_stats  # 返回统计结果

# 假设 df_1d, df_5min, df_15min 已经定义
# 对日线数据进行分析
stats_1d = analyze_price_crossing(df_1d, "1d", price_step=10)
print("\n日线数据 (1d) 穿越统计结果:")
print(stats_1d)

# 对5分钟数据进行分析，步长设置为 10
stats_5min = analyze_price_crossing(df_5min, "5min", price_step=10)
print("\n5分钟数据 (5min) 穿越统计结果:")
print(stats_5min)

# 对15分钟数据进行分析，步长设置为 10
stats_15min = analyze_price_crossing(df_15min, "15min", price_step=10)
print("\n15分钟数据 (15min) 穿越统计结果:")
print(stats_15min)
```

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import matplotlib.pyplot as plt  # 导入 matplotlib.pyplot 库，用于绘图
import matplotlib.ticker as ticker  # 导入 matplotlib.ticker 库，用于设置刻度
import numpy as np  # 导入 numpy 库，用于数值计算

# 加载数据 (代码与之前相同，此处省略)
df_1d = pd.read_csv('hstech_data_1d.csv')  # 读取日线数据
df_5min = pd.read_csv('hstech_data_5min.csv')  # 读取 5 分钟数据
df_15min = pd.read_csv('hstech_data_15min.csv')  # 读取 15 分钟数据

df_1d['date'] = pd.to_datetime(df_1d['date'])  # 将日线数据的日期列转换为 datetime 类型
df_5min['date'] = pd.to_datetime(df_5min['date'])  # 将 5 分钟数据的日期列转换为 datetime 类型
df_15min['date'] = pd.to_datetime(df_15min['date'])  # 将 15 分钟数据的日期列转换为 datetime 类型

print("日线数据 (1d) 示例:")  # 打印日线数据示例
print(df_1d.head())  # 打印日线数据的前几行
print("\n5分钟数据 (5min) 示例:")  # 打印 5 分钟数据示例
print(df_5min.head())  # 打印 5 分钟数据的前几行
print("\n15分钟数据 (15min) 示例:")  # 打印 15 分钟数据示例
print(df_15min.head())  # 打印 15 分钟数据的前几行


def analyze_price_crossing(df, time_frame_name):
    """
    分析价格穿越次数并绘图，使用双Y轴，优化刻度，自动设置价格范围和步长，棋盘格背景。

    参数:
    df: DataFrame, 输入数据
    time_frame_name: str, 时间周期名称 (例如 "1d", "5min", "15min")
    """

    # 1. 确定价格范围和步长 (代码与之前相同，此处省略)
    min_price = df[['open', 'high', 'low', 'close']].min().min()  # 获取最低价
    max_price = df[['open', 'high', 'low', 'close']].max().max()  # 获取最高价

    start_price = min_price * 0.98  # 设置起始价格，向下扩展 2%
    end_price = max_price * 1.02  # 设置结束价格，向上扩展 2%

    price_range_value = max_price - min_price  # 计算价格范围
    if time_frame_name == '1d':  # 如果是日线数据
        price_step = price_range_value / 200  # 设置价格步长
    elif time_frame_name == '15min':  # 如果是 15 分钟数据
        price_step = price_range_value / 300  # 设置价格步长
    elif time_frame_name == '5min':  # 如果是 5 分钟数据
        price_step = price_range_value / 400  # 设置价格步长
    else:  # 如果是其他时间周期
        price_step = price_range_value / 100  # 设置价格步长
    price_step = max(price_step, 0.01)  # 确保价格步长不小于 0.01

    price_range = []  # 初始化价格范围列表
    current_price = start_price  # 设置当前价格为起始价格
    while current_price <= end_price:  # 循环生成价格范围
        price_range.append(current_price)  # 将当前价格添加到价格范围列表
        current_price += price_step  # 更新当前价格

    crossing_stats = {}  # 初始化存储结果的字典

    # 2. 遍历价格 (代码与之前相同，此处省略)
    for price_level in price_range:  # 遍历价格范围
        crossing_stats[price_level] = {'crossing_times': 0, 'volume_before_crossing': 0, 'volume_after_crossing': 0}  # 初始化当前价格水平的统计信息
        daily_data = df.groupby(df['date'].dt.date)  # 按天分组

        for date, daily_df in daily_data:  # 遍历每天的数据
            daily_df = daily_df.sort_values(by='date')  # 确保日内数据按时间排序
            last_close = None  # 初始化上一个收盘价
            crossing_type = None  # 记录上一次穿越类型 (up or down)
            daily_volume_before = 0  # 初始化当日穿越前成交量
            daily_volume_after = 0  # 初始化当日穿越后成交量
            volume_before_flag = True  # 标记是否在统计穿越前的成交量

            for index, row in daily_df.iterrows():  # 遍历日内每条数据
                current_close = row['close']  # 获取当前收盘价
                current_volume = row['volume']  # 获取当前成交量

                if last_close is not None:  # 如果存在上一个收盘价
                    up_crossing = (last_close <= price_level) and (current_close > price_level)  # 判断是否向上穿越
                    down_crossing = (last_close >= price_level) and (current_close < price_level)  # 判断是否向下穿越

                    if up_crossing:  # 如果向上穿越
                        if crossing_type == 'down':  # 如果上一次是向下穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'up'  # 更新穿越类型为向上
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                    elif down_crossing:  # 如果向下穿越
                        if crossing_type == 'up':  # 如果上一次是向上穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'down'  # 更新穿越类型为向下
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                if volume_before_flag:  # 如果在统计穿越前的成交量
                    daily_volume_before += current_volume  # 累加穿越前的成交量

                last_close = current_close  # 更新上一个收盘价

    # 5. 结果可视化 (修改后的绘图部分)
    prices = list(crossing_stats.keys())  # 获取价格列表
    crossing_counts = [crossing_stats[p]['crossing_times'] for p in prices]  # 获取穿越次数列表
    volume_before = [crossing_stats[p]['volume_before_crossing'] for p in prices]  # 获取穿越前累计成交量列表
    volume_after = [crossing_stats[p]['volume_after_crossing'] for p in prices]  # 获取穿越后累计成交量列表
    total_volume = [v_before + v_after for v_before, v_after in zip(volume_before, volume_after)]  # 计算总累计成交量

    fig, ax1 = plt.subplots(figsize=(12, 6))  # 创建图形对象，设置尺寸

    # 左轴：累计成交量
    ax1.plot(prices, total_volume, label='累计成交量', color='blue')  # 绘制累计成交量折线图
    ax1.set_xlabel('价格水平')  # 设置 x 轴标签
    ax1.set_ylabel('累计成交量', color='blue')  # 设置 y 轴标签
    ax1.tick_params(axis='y', labelcolor='blue')  # 设置 y 轴刻度标签颜色为蓝色

    # 右轴：穿越次数
    ax2 = ax1.twinx()  # 创建共享 x 轴的第二个 y 轴
    ax2.plot(prices, crossing_counts, label='穿越次数', color='red')  # 绘制穿越次数折线图
    ax2.set_ylabel('穿越次数', color='red')  # 设置 y 轴标签
    ax2.tick_params(axis='y', labelcolor='red')  # 设置 y 轴刻度标签颜色为红色

    # 优化刻度显示 - X轴更密集，Y轴更稀疏
    ax1.xaxis.set_major_locator(ticker.MultipleLocator(base=calculate_optimal_locator_base(prices) / 2))  # 设置 x 轴主要刻度定位器，使其更密集
    ax1.yaxis.set_major_locator(ticker.MultipleLocator(base=calculate_optimal_locator_base(total_volume) * 2))  # 设置 y 轴主要刻度定位器，使其更稀疏
    ax2.yaxis.set_major_locator(ticker.MultipleLocator(base=calculate_optimal_locator_base(crossing_counts) * 2))  # 设置 y 轴主要刻度定位器，使其更稀疏

    plt.title(f'{time_frame_name} - 不同价格水平的反复穿越次数与累计成交量')  # 设置标题
    fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.95))  # 设置图例位置

    # 棋盘格背景
    ax1.set_facecolor('#f0f0f0')  # 设置 Axes 背景色为浅灰色
    ax1.xaxis.set_minor_locator(ticker.AutoMinorLocator(n=2))  # X轴设置次要刻度， AutoMinorLocator自动选择次要刻度数量
    ax1.yaxis.set_minor_locator(ticker.AutoMinorLocator(n=2))  # Y轴设置次要刻度
    ax1.grid(which='major', axis='both', linestyle='-', linewidth='0.5', color='gray')  # 主要网格线
    ax1.grid(which='minor', axis='both', linestyle=':', linewidth='0.5', color='lightgray')  # 次要网格线，linestyle=':' 虚线，更像棋盘格

    plt.tight_layout()  # 调整子图参数，使之填充整个图像区域
    plt.show()  # 显示图形
    # [Image of Line chart with chessboard grid, refined X-axis ticks and coarser Y-axis ticks showing volume and crossing counts at different price levels for {time_frame_name} data]

    return crossing_stats  # 返回统计结果


def calculate_optimal_locator_base(data_list):
    """
    自适应计算 MultipleLocator 的 base 值，用于优化刻度密度。
    """
    if not data_list:  # 如果数据列表为空
        return 1  # 返回 1，避免空列表
    data_range = max(data_list) - min(data_list)  # 计算数据范围
    if data_range == 0:  # 如果数据范围为 0
        return 1  # 返回 1，避免除以 0
    order_of_magnitude = 10**int(np.log10(data_range)) if data_range > 0 else 1  # 计算数量级
    return order_of_magnitude / 5 if order_of_magnitude >= 10 else order_of_magnitude / 2  # 返回优化的刻度基数


# 对日线数据进行分析
stats_1d = analyze_price_crossing(df_1d, "1d")  # 分析日线数据
print("\n日线数据 (1d) 穿越统计结果:")  # 打印日线数据统计结果
print(stats_1d)  # 打印日线数据统计结果

# 对5分钟数据进行分析
stats_5min = analyze_price_crossing(df_5min, "5min")  # 分析 5 分钟数据
print("\n5分钟数据 (5min) 穿越统计结果:")  # 打印 5 分钟数据统计结果
print(stats_5min)  # 打印 5 分钟数据统计结果

# 对15分钟数据进行分析
stats_15min = analyze_price_crossing(df_15min, "15min")  # 分析 15 分钟数据
print("\n15分钟数据 (15min) 穿越统计结果:")  # 打印 15 分钟数据统计结果
print(stats_15min)  # 打印 15 分钟数据统计结果
```

当然！我们来深入分析一下穿越次数和穿越前后累计成交量所代表的含义。理解这些指标可以帮助我们更好地解读图表，并可能从中获取一些交易上的启发。

**1. 穿越次数 (Crossing Times) 的含义**

*   **定义:**  穿越次数，指的是在**同一天内**，价格围绕着特定价格水平线，发生**先穿过再反向穿回**的往返次数。例如，价格先从下方上穿价格线，然后在当天又从上方下穿回来，或者反过来，都算作一次穿越。

*   **市场解读:**  高穿越次数的价格水平，通常意味着：

    *   **价格在该水平附近震荡频繁:**  穿越次数越多，说明价格在这个特定价格附近来回波动的次数就越多。这暗示市场在该价格点位上多空双方存在显著的分歧，价格方向不明朗，呈现出**拉锯战**的状态。

    *   **潜在的阻力或支撑区域:**  反复穿越可能表明该价格水平附近存在一定的**阻力**或**支撑**。当价格接近这些水平时，买方和卖方力量在此处交汇，导致价格难以有效突破，转而呈现震荡或反转。

    *   **市场在该价格水平上犹豫不决:**  高穿越次数也可能反映市场对该价格水平的价值判断存在**犹豫**。多空双方力量相对均衡，无法形成单边行情，导致价格围绕该水平上下波动。

    *   **波动性可能较高:**  在穿越次数较多的价格区域，价格波动性往往相对较高。频繁的往返穿越本身就体现了价格的剧烈变化。

*   **图表上的体现:** 在您生成的图表中，穿越次数以折线的形式展示，数值越高，折线越高。  **峰值** 就代表穿越次数最多的价格水平，这些价格水平就值得我们重点关注。

**2. 穿越前后累计成交量 (Cumulative Volume Before & After Crossing) 的含义**

*   **定义:**
    *   **穿越前累计成交量:**  指的是在**价格首次有效穿越**某个价格水平线**之前**，当天累积的成交量总和。
    *   **穿越后累计成交量:**  指的是在**价格首次有效穿越**某个价格水平线**之后**，当天剩余时间内累积的成交量总和。

*   **市场解读:**  穿越前后成交量的对比，以及它们与穿越次数的结合，可以提供更丰富的市场信息：

    *   **成交量验证突破的有效性:**
        *   **穿越后成交量显著放大 (相对于穿越前成交量):**  如果价格突破某个高穿越次数的价格水平后，成交量明显放大，可能意味着这次突破是**相对有效**的。放大的成交量可以被解读为市场参与者对新方向的**认可**和**追随**，突破后价格可能更倾向于延续趋势。
        *   **穿越后成交量没有放大甚至萎缩:**  如果价格突破后，成交量没有明显变化，甚至反而萎缩，那么这次突破可能就**值得怀疑**。这可能是一个**虚假突破**信号，价格可能很快会反转回原先的震荡区间。

    *   **成交量印证价格震荡区间的强度:**
        *   **在穿越次数高的价格水平，穿越前后成交量都很大:**  这可能进一步印证了该价格水平作为重要**支撑或阻力**的强度。多空双方在该位置的争夺非常激烈，投入了大量的资金（成交量大），但仍然难以形成明确的突破方向，导致价格反复穿越。

    *   **成交量与价格行为的背离:**
        *   **高穿越次数，但穿越前后成交量都很低:**  这可能暗示虽然价格在某个水平附近震荡，但市场参与度不高，可能只是**小范围资金**在试探，或者市场整体交投清淡。这种情况下，价格的震荡可能不具有太强的趋势指示意义。

*   **图表上的体现:** 在您的图表中，穿越前后累计成交量分别用绿色和红色折线表示。

    *   **绿色线 (穿越前累计成交量):**  可以理解为价格在试探某个价格水平时的“蓄力”情况。
    *   **红色线 (穿越后累计成交量):**  可以理解为价格在突破（或跌破）某个价格水平后的“爆发”情况。

    对比绿色和红色线的相对高度，以及它们在穿越次数峰值区域的表现，可以帮助我们判断不同价格水平的市场强度和潜在的突破有效性。

**综合应用： 如何解读图表并辅助交易决策**

1.  **寻找“穿越次数峰值”区域:**  首先，在“穿越次数”图表中，找出穿越次数最高的那些价格水平。这些水平很可能是市场关注的焦点。

2.  **观察“成交量”图表在峰值区域的表现:**  然后，结合“累计成交量”图表，观察在穿越次数峰值对应的价格水平上，穿越前后的成交量情况。

    *   **高穿越次数 + 穿越后成交量放大:**  这可能暗示价格在该水平附近震荡后，一旦突破，可能会形成较明显的趋势行情。可以关注突破方向，结合其他指标判断是否顺势交易。
    *   **高穿越次数 + 穿越后成交量没有放大 (或很小):**  这可能暗示突破的有效性不足，价格可能很快回到震荡区间。  此时，如果价格回到震荡区间，可以考虑在该区间内进行高抛低吸的区间交易策略。
    *   **高穿越次数 + 穿越前后成交量都很大:**  这可能表明该价格水平是重要的多空分水岭，市场在该位置存在显著分歧。 交易上需要谨慎，最好等待更明确的方向信号出现，或者避免在该区域附近进行激进的突破交易。
    *   **低穿越次数 + 成交量平淡:**  表明该价格水平可能不是市场关注的重点，价格穿越可能比较随意，不具备太强的交易参考价值。

3.  **结合不同时间周期分析:**  您可以对比分析日线、5分钟线、15分钟线等不同时间周期的图表结果。

    *   **日线图表:**  更侧重于分析**中长期的关键价格水平**，判断大趋势和主要支撑阻力区域。
    *   **分钟线图表:**  更侧重于分析**短线交易机会**，例如日内交易的入场点和离场点，以及短线波动较为剧烈的价格区域。

4.  **结合其他技术指标和基本面分析:**  请记住，任何单一的分析方法都有局限性。“穿越次数和成交量分析”可以作为您交易决策的**辅助工具**，最好结合其他技术指标（例如均线、MACD、RSI等）和基本面信息进行综合判断，以提高交易决策的准确性。

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import plotly.graph_objects as go  # 导入 plotly.graph_objects 库，用于创建图形对象
from plotly.subplots import make_subplots  # 导入 plotly.subplots 库，用于创建子图
import numpy as np  # 导入 numpy 库，用于数值计算

# 加载数据 (保持与之前代码一致)
df_1d = pd.read_csv('hstech_data_1d.csv')  # 读取日线数据
df_5min = pd.read_csv('hstech_data_5min.csv')  # 读取 5 分钟数据
df_15min = pd.read_csv('hstech_data_15min.csv')  # 读取 15 分钟数据

df_1d['date'] = pd.to_datetime(df_1d['date'])  # 将日线数据的日期列转换为 datetime 类型
df_5min['date'] = pd.to_datetime(df_5min['date'])  # 将 5 分钟数据的日期列转换为 datetime 类型
df_15min['date'] = pd.to_datetime(df_15min['date'])  # 将 15 分钟数据的日期列转换为 datetime 类型

print("日线数据 (1d) 示例:")  # 打印日线数据示例
print(df_1d.head())  # 打印日线数据的前几行
print("\n5分钟数据 (5min) 示例:")  # 打印 5 分钟数据示例
print(df_5min.head())  # 打印 5 分钟数据的前几行
print("\n15分钟数据 (15min) 示例:")  # 打印 15 分钟数据示例
print(df_15min.head())  # 打印 15 分钟数据的前几行


def analyze_price_crossing(df, time_frame_name):
    """
    分析价格穿越次数并使用 Plotly 绘制交互式图表，双Y轴，优化刻度，自动设置价格范围和步长，棋盘格背景，交互式显示数据点信息。

    参数:
    df: DataFrame, 输入数据
    time_frame_name: str, 时间周期名称 (例如 "1d", "5min", "15min")
    """

    # 1. 确定价格范围和步长 (保持与之前代码一致)
    min_price = df[['open', 'high', 'low', 'close']].min().min()  # 获取最低价
    max_price = df[['open', 'high', 'low', 'close']].max().max()  # 获取最高价

    start_price = min_price * 0.98  # 设置起始价格，向下扩展 2%
    end_price = max_price * 1.02  # 设置结束价格，向上扩展 2%

    price_range_value = max_price - min_price  # 计算价格范围
    if time_frame_name == '1d':  # 如果是日线数据
        price_step = price_range_value / 50  # 设置价格步长
    elif time_frame_name == '15min':  # 如果是 15 分钟数据
        price_step = price_range_value / 200  # 设置价格步长
    elif time_frame_name == '5min':  # 如果是 5 分钟数据
        price_step = price_range_value / 400  # 设置价格步长
    else:  # 如果是其他时间周期
        price_step = price_range_value / 100  # 设置价格步长
    price_step = max(price_step, 0.01)  # 确保价格步长不小于 0.01

    price_range = []  # 初始化价格范围列表
    current_price = start_price  # 设置当前价格为起始价格
    while current_price <= end_price:  # 循环生成价格范围
        price_range.append(current_price)  # 将当前价格添加到价格范围列表
        current_price += price_step  # 更新当前价格

    crossing_stats = {}  # 初始化存储结果的字典

    # 2. 遍历价格 (保持与之前代码一致)
    for price_level in price_range:  # 遍历价格范围
        crossing_stats[price_level] = {'crossing_times': 0, 'volume_before_crossing': 0, 'volume_after_crossing': 0}  # 初始化当前价格水平的统计信息
        daily_data = df.groupby(df['date'].dt.date)  # 按天分组

        for date, daily_df in daily_data:  # 遍历每天的数据
            daily_df = daily_df.sort_values(by='date')  # 确保日内数据按时间排序
            last_close = None  # 初始化上一个收盘价
            crossing_type = None  # 记录上一次穿越类型 (up or down)
            daily_volume_before = 0  # 初始化当日穿越前成交量
            daily_volume_after = 0  # 初始化当日穿越后成交量
            volume_before_flag = True  # 标记是否在统计穿越前的成交量

            for index, row in daily_df.iterrows():  # 遍历日内每条数据
                current_close = row['close']  # 获取当前收盘价
                current_volume = row['volume']  # 获取当前成交量

                if last_close is not None:  # 如果存在上一个收盘价
                    up_crossing = (last_close <= price_level) and (current_close > price_level)  # 判断是否向上穿越
                    down_crossing = (last_close >= price_level) and (current_close < price_level)  # 判断是否向下穿越

                    if up_crossing:  # 如果向上穿越
                        if crossing_type == 'down':  # 如果上一次是向下穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'up'  # 更新穿越类型为向上
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                    elif down_crossing:  # 如果向下穿越
                        if crossing_type == 'up':  # 如果上一次是向上穿越
                            crossing_stats[price_level]['crossing_times'] += 1  # 穿越次数加 1
                            crossing_stats[price_level]['volume_before_crossing'] += daily_volume_before  # 累加穿越前成交量
                            crossing_stats[price_level]['volume_after_crossing'] += (daily_df['volume'].sum() - daily_volume_before)  # 累加穿越后成交量
                        crossing_type = 'down'  # 更新穿越类型为向下
                        volume_before_flag = False  # 开始统计穿越后的成交量
                        daily_volume_after = 0  # 重置当日穿越后成交量

                if volume_before_flag:  # 如果在统计穿越前的成交量
                    daily_volume_before += current_volume  # 累加穿越前的成交量

                last_close = current_close  # 更新上一个收盘价

    # 5. 结果可视化 (使用 Plotly 绘制交互式图表)
    prices = list(crossing_stats.keys())  # 获取价格列表
    crossing_counts = [crossing_stats[p]['crossing_times'] for p in prices]  # 获取穿越次数列表
    volume_before = [crossing_stats[p]['volume_before_crossing'] for p in prices]  # 获取穿越前累计成交量列表
    volume_after = [crossing_stats[p]['volume_after_crossing'] for p in prices]  # 获取穿越后累计成交量列表
    total_volume = [v_before + v_after for v_before, v_after in zip(volume_before, volume_after)]  # 计算总累计成交量

    fig = make_subplots(specs=[[{"secondary_y": True}]])  # 创建带次坐标轴的图表

    # 左轴：累计成交量
    fig.add_trace(go.Scatter(x=prices, y=total_volume, name='累计成交量', line=dict(color='blue')), secondary_y=False)  # 添加累计成交量折线图

    # 右轴：穿越次数
    fig.add_trace(go.Scatter(x=prices, y=crossing_counts, name='穿越次数', line=dict(color='red')), secondary_y=True)  # 添加穿越次数折线图

    # 设置布局
    fig.update_layout(
        title=f'{time_frame_name} - 不同价格水平的反复穿越次数与累计成交量 (交互式)',  # 设置标题
        xaxis_title='价格水平',  # 设置 x 轴标题
        yaxis_title='累计成交量',  # 设置左轴标题
        yaxis2_title='穿越次数',  # 设置右轴标题
        yaxis=dict(
            gridcolor='#e0e0e0',  # 主网格线颜色
            zerolinecolor='black',  # 零线颜色
            tickformat=',.0f',  # Y轴刻度格式，不显示小数点
            showgrid=True,  # 显示主网格线
        ),
        yaxis2=dict(
            gridcolor='#e0e0e0',  # 次坐标轴网格线颜色
            zerolinecolor='black',  # 零线颜色
            tickformat=',.0f',  # Y轴刻度格式，不显示小数点
            showgrid=True,  # 显示次坐标轴网格线
            overlaying='y',  # 关键: 共享X轴
            side='right'  # 次坐标轴显示在右侧
        ),
        xaxis=dict(
            gridcolor='#e0e0e0',  # X轴网格线颜色
            zerolinecolor='black',  # 零线颜色
            tickformat=',.2f',  # X轴刻度格式，保留两位小数
            showgrid=True,  # 显示X轴网格线
        ),
        plot_bgcolor='#f0f0f0',  # 图表背景色
        showlegend=True,  # 显示图例
        hovermode="closest"  # 鼠标悬停模式，closest表示显示最近的数据点信息
    )

    # 添加棋盘格背景 (通过修改网格线样式实现类似效果)
    fig.update_xaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True)  # X轴次网格线
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=False)  # 左Y轴次网格线
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=True)  # 右Y轴次网格线

    fig.show()  # 显示图形
    # [Image of Interactive Line chart with chessboard grid, refined X-axis ticks and coarser Y-axis ticks showing volume and crossing counts at different price levels for {time_frame_name} data, with hover information in Plotly]

    return crossing_stats  # 返回统计结果


def calculate_optimal_locator_base(data_list):  # 保留该函数，但Plotly刻度自适应性更强，实际未使用到这个函数优化刻度
    """
    自适应计算 MultipleLocator 的 base 值， 用于优化刻度密度 (该函数在 Plotly 版本中可能不再直接使用).
    """
    if not data_list:  # 如果数据列表为空
        return 1  # 返回 1，避免空列表
    data_range = max(data_list) - min(data_list)  # 计算数据范围
    if data_range == 0:  # 如果数据范围为 0
        return 1  # 返回 1，避免除以 0
    order_of_magnitude = 10**int(np.log10(data_range)) if data_range > 0 else 1  # 计算数量级
    return order_of_magnitude / 5 if order_of_magnitude >= 10 else order_of_magnitude / 2  # 返回优化的刻度基数


# 对日线数据进行分析
stats_1d = analyze_price_crossing(df_1d, "1d")  # 分析日线数据
print("\n日线数据 (1d) 穿越统计结果:")  # 打印日线数据统计结果
print(stats_1d)  # 打印日线数据统计结果

# 对5分钟数据进行分析
stats_5min = analyze_price_crossing(df_5min, "5min")  # 分析 5 分钟数据
print("\n5分钟数据 (5min) 穿越统计结果:")  # 打印 5 分钟数据统计结果
print(stats_5min)  # 打印 5 分钟数据统计结果

# 对15分钟数据进行分析
stats_15min = analyze_price_crossing(df_15min, "15min")  # 分析 15 分钟数据
print("\n15分钟数据 (15min) 穿越统计结果:")  # 打印 15 分钟数据统计结果
print(stats_15min)  # 打印 15 分钟数据统计结果

```

核心思路：区分向上穿越和向下穿越，并分别统计相关指标。

目前的代码，只是简单统计了价格围绕某个水平线“穿越”的次数，以及穿越前后的成交量。但是，一次“穿越”其实包含两种方向：

向上穿越 (Upward Crossing): 价格从下方突破价格水平线，运行到上方。这通常被视为一种偏多信号，代表买方力量增强。
向下穿越 (Downward Crossing): 价格从上方跌破价格水平线，运行到下方。这通常被视为一种偏空信号，代表卖方力量增强。


穿越次数图表:  可以将 向上穿越次数 和 向下穿越次数 分别用两条不同的折线展示。例如，向上穿越用绿色（代表看涨），向下穿越用红色（代表看跌）。

成交量图表:  同样，可以将 向上穿越时累计成交量 和 向下穿越时累计成交量 分别用两条不同的折线展示，颜色也与穿越次数图表保持一致（绿色代表向上，红色代表向下）。

趋势研判思路:

向上穿越次数 和 向下穿越次数 的对比：

向上穿越次数显著多于向下穿越次数： 可能暗示在该价格水平附近，多方力量更强，价格未来可能更倾向于向上突破发展。
向下穿越次数显著多于向上穿越次数： 可能暗示在该价格水平附近，空方力量更强，价格未来可能更倾向于向下突破发展。
向上和向下穿越次数接近： 可能暗示多空力量均衡，价格在该水平附近可能继续震荡。
向上穿越成交量 和 向下穿越成交量 的对比：

向上穿越成交量显著大于向下穿越成交量： 可能进一步验证向上突破的强度，暗示向上突破可能更有效，后续上涨趋势可能更强劲。
向下穿越成交量显著大于向上穿越成交量： 可能进一步验证向下突破的强度，暗示向下突破可能更有效，后续下跌趋势可能更强劲。
向上和向下穿越成交量接近： 可能暗示突破方向不明朗，需要结合穿越次数以及其他指标综合判断。

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import plotly.graph_objects as go  # 导入 plotly.graph_objects 库，用于创建图形对象
from plotly.subplots import make_subplots  # 导入 plotly.subplots 库，用于创建子图
import numpy as np  # 导入 numpy 库，用于数值计算

# 加载数据 (保持与之前代码一致)
df_1d = pd.read_csv('hstech_data_1d.csv')  # 读取日线数据
df_5min = pd.read_csv('hstech_data_5min.csv')  # 读取 5 分钟数据
df_15min = pd.read_csv('hstech_data_15min.csv')  # 读取 15 分钟数据

df_1d['date'] = pd.to_datetime(df_1d['date'])  # 将日线数据的日期列转换为 datetime 类型
df_5min['date'] = pd.to_datetime(df_5min['date'])  # 将 5 分钟数据的日期列转换为 datetime 类型
df_15min['date'] = pd.to_datetime(df_15min['date'])  # 将 15 分钟数据的日期列转换为 datetime 类型

print("日线数据 (1d) 示例:")  # 打印日线数据示例
print(df_1d.head())  # 打印日线数据的前几行
print("\n5分钟数据 (5min) 示例:")  # 打印 5 分钟数据示例
print(df_5min.head())  # 打印 5 分钟数据的前几行
print("\n15分钟数据 (15min) 示例:")  # 打印 15 分钟数据示例
print(df_15min.head())  # 打印 15 分钟数据的前几行


def analyze_price_crossing_directional(df, time_frame_name):
    """
    分析价格的向上和向下穿越次数及成交量，并使用 Plotly 绘制分方向的交互式图表。
    左轴显示柱状的穿越成交量（区分向上/下），右轴显示折线穿越次数（区分向上/下）。

    参数:
    df: DataFrame, 输入数据
    time_frame_name: str, 时间周期名称 (例如 "1d", "5min", "15min")
    """

    # 1. 确定价格范围和步长 (保持与之前代码一致)
    min_price = df[['open', 'high', 'low', 'close']].min().min()  # 获取最低价
    max_price = df[['open', 'high', 'low', 'close']].max().max()  # 获取最高价

    start_price = min_price * 0.98  # 设置起始价格，向下扩展 2%
    end_price = max_price * 1.02  # 设置结束价格，向上扩展 2%

    price_range_value = max_price - min_price  # 计算价格范围
    if time_frame_name == '1d':  # 如果是日线数据
        price_step = price_range_value / 50  # 设置价格步长
    elif time_frame_name == '15min':  # 如果是 15 分钟数据
        price_step = price_range_value / 200  # 设置价格步长
    elif time_frame_name == '5min':  # 如果是 5 分钟数据
        price_step = price_range_value / 400  # 设置价格步长
    else:  # 如果是其他时间周期
        price_step = price_range_value / 100  # 设置价格步长
    price_step = max(price_step, 0.01)  # 确保价格步长不小于 0.01

    price_range = []  # 初始化价格范围列表
    current_price = start_price  # 设置当前价格为起始价格
    while current_price <= end_price:  # 循环生成价格范围
        price_range.append(current_price)  # 将当前价格添加到价格范围列表
        current_price += price_step  # 更新当前价格

    directional_crossing_stats = {}  # 存储分方向穿越统计结果的字典

    # 2. 遍历价格，统计分方向穿越次数和穿越成交量 (代码与之前相同)
    for price_level in price_range:  # 遍历价格范围
        directional_crossing_stats[price_level] = {'up_crossing_times': 0, 'down_crossing_times': 0, 'up_crossing_volume': 0, 'down_crossing_volume': 0}  # 初始化当前价格水平的分方向统计信息
        daily_data = df.groupby(df['date'].dt.date)  # 按天分组

        for date, daily_df in daily_data:  # 遍历每天的数据
            daily_df = daily_df.sort_values(by='date')  # 确保日内数据按时间排序
            last_close = None  # 初始化上一个收盘价

            for index, row in daily_df.iterrows():  # 遍历日内每条数据
                current_close = row['close']  # 获取当前收盘价
                current_volume = row['volume']  # 获取当前成交量

                if last_close is not None:  # 如果存在上一个收盘价
                    up_crossing = (last_close <= price_level) and (current_close > price_level)  # 判断是否向上穿越
                    down_crossing = (last_close >= price_level) and (current_close < price_level)  # 判断是否向下穿越

                    if up_crossing:  # 如果向上穿越
                        directional_crossing_stats[price_level]['up_crossing_times'] += 1  # 向上穿越次数加 1
                        directional_crossing_stats[price_level]['up_crossing_volume'] += current_volume  # 累加向上穿越成交量
                    elif down_crossing:  # 如果向下穿越
                        directional_crossing_stats[price_level]['down_crossing_times'] += 1  # 向下穿越次数加 1
                        directional_crossing_stats[price_level]['down_crossing_volume'] += current_volume  # 累加向下穿越成交量

                last_close = current_close  # 更新上一个收盘价

    # 5. 结果可视化 (使用 Plotly 绘制分方向的交互式图表 - 使用柱状图和折线图)
    prices = list(directional_crossing_stats.keys())  # 获取价格列表
    up_crossing_counts = [directional_crossing_stats[p]['up_crossing_times'] for p in prices]  # 获取向上穿越次数列表
    down_crossing_counts = [directional_crossing_stats[p]['down_crossing_times'] for p in prices]  # 获取向下穿越次数列表
    up_crossing_volumes = [directional_crossing_stats[p]['up_crossing_volume'] for p in prices]  # 获取向上穿越成交量列表
    down_crossing_volumes = [directional_crossing_stats[p]['down_crossing_volume'] for p in prices]  # 获取向下穿越成交量列表

    fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])  # 创建带次坐标轴的图表

    # 左轴：柱状图 - 向上穿越成交量 (浅绿色)
    fig.add_trace(go.Bar(x=prices, y=up_crossing_volumes, name='向上穿越成交量', marker_color='lightgreen',
                             hovertemplate="价格水平: %{x:.2f}<br>向上穿越成交量: %{y:.2f}",  # 柱状图 hovertemplate
                             ),
                  secondary_y=False, row=1, col=1)  # 浅绿色柱状图

    # 左轴：柱状图 - 向下穿越成交量 (浅红色)
    fig.add_trace(go.Bar(x=prices, y=down_crossing_volumes, name='向下穿越成交量', marker_color='lightcoral',
                             hovertemplate="价格水平: %{x:.2f}<br>向下穿越成交量: %{y:.2f}",  # 柱状图的 hovertemplate
                             ),
                  secondary_y=False, row=1, col=1)  # 浅红色柱状图

    # 右轴：折线图 - 向上穿越次数 (深绿色)
    fig.add_trace(go.Scatter(x=prices, y=up_crossing_counts, name='向上穿越次数', line=dict(color='green'),  # 添加向上穿越次数的折线图
                             hovertemplate="价格水平: %{x:.2f}<br>向上穿越次数: %{y}",  # 设置折线图的悬停提示信息
                             ),
                  secondary_y=True, row=1, col=1)  # 将折线图添加到次坐标轴，深绿色

    # 右轴：折线图 - 向下穿越次数 (深红色)
    fig.add_trace(go.Scatter(x=prices, y=down_crossing_counts, name='向下穿越次数', line=dict(color='red'),  # 添加向下穿越次数的折线图
                             hovertemplate="价格水平: %{x:.2f}<br>向下穿越次数: %{y}",  # 设置折线图的悬停提示信息
                             ),
                  secondary_y=True, row=1, col=1)  # 将折线图添加到次坐标轴，深红色


    # 设置布局
    fig.update_layout(
        title=f'{time_frame_name} - 不同价格水平的向上/下穿越次数和成交量 (柱状图-成交量, 折线图-次数)',  # 设置图表标题，包含时间周期名称
        xaxis_title='价格水平',  # 设置 x 轴标题
        yaxis_title='穿越成交量',  # 设置 y 轴标题（左轴），修改为 "穿越成交量"
        yaxis2_title='穿越次数',  # 设置 y2 轴标题（右轴），保持 "穿越次数"
        yaxis=dict(  # 设置 y 轴（左轴）样式
            title_text="穿越成交量",  # 设置 y 轴标题文本
            title_font=dict(color='black'),  # 设置 y 轴标题字体颜色
            tickfont=dict(color='black'),  # 设置 y 轴刻度字体颜色
            gridcolor='#e0e0e0',  # 设置 y 轴网格线颜色
            zerolinecolor='black',  # 设置 y 轴零线颜色
            tickformat=',.2f',  # 设置 y 轴刻度格式，保留两位小数
            showgrid=True,  # 显示 y 轴网格线
        ),
        yaxis2=dict(  # 设置 y2 轴（右轴）样式
            title_text="穿越次数",  # 设置 y2 轴标题文本
            title_font=dict(color='black'),  # 设置 y2 轴标题字体颜色
            tickfont=dict(color='black'),  # 设置 y2 轴刻度字体颜色
            gridcolor='#e0e0e0',  # 设置 y2 轴网格线颜色
            zerolinecolor='black',  # 设置 y2 轴零线颜色
            tickformat=',.0f',  # 设置 y2 轴刻度格式，整数
            showgrid=True,  # 显示 y2 轴网格线
            overlaying='y',  # 将 y2 轴覆盖在 y 轴上
            side='right'  # 设置 y2 轴在右侧显示
        ),
        xaxis=dict(  # 设置 x 轴样式
            gridcolor='#e0e0e0',  # 设置 x 轴网格线颜色
            zerolinecolor='black',  # 设置 x 轴零线颜色
            tickformat=',.2f',  # 设置 x 轴刻度格式，保留两位小数
            showgrid=True,  # 显示 x 轴网格线
        ),
        plot_bgcolor='#f0f0f0',  # 设置图表背景颜色
        showlegend=True,  # 显示图例
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),  # 设置图例水平显示在图表底部
        hovermode="closest"  # 设置悬停模式为最近点
    )

    # 添加棋盘格背景 (通过修改网格线样式实现类似效果)
    fig.update_xaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True)  # 设置 x 轴次要刻度样式，实现棋盘格效果
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=False)  # 设置 y 轴（左轴）次要刻度样式，实现棋盘格效果
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=True)  # 设置 y2 轴（右轴）次要刻度样式，实现棋盘格效果


    fig.show()  # 显示图表
    # [Image of Interactive chart with chessboard grid, refined X-axis ticks and coarser Y-axis ticks showing directional crossing counts (lines) and crossing volume (bars - light color) at different price levels for {time_frame_name} data, with enhanced hover information in Plotly, now with Volume on the left Y-axis and Counts on the right Y-axis]  # 图表描述（注释）


    return directional_crossing_stats  # 返回分方向穿越统计结果


def calculate_optimal_locator_base(data_list):  # 定义计算最优刻度基数的函数
    """
    自适应计算 MultipleLocator 的 base 值， 用于优化刻度密度 (该函数在 Plotly 版本中可能不再直接使用).
    """
    if not data_list:  # 如果数据列表为空
        return 1  # 返回 1
    data_range = max(data_list) - min(data_list)  # 计算数据范围
    if data_range == 0:  # 如果数据范围为 0
        return 1  # 返回 1
    order_of_magnitude = 10**int(np.log10(data_range)) if data_range > 0 else 1  # 计算数量级
    return order_of_magnitude / 5 if order_of_magnitude >= 10 else order_of_magnitude / 2  # 返回最优刻度基数


# 对日线数据进行分析 (函数名修改为 directional 版本)
stats_1d_directional = analyze_price_crossing_directional(df_1d, "1d")  # 分析日线数据
print("\n日线数据 (1d) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印日线数据分析结果提示信息
print(stats_1d_directional)  # 打印日线数据分析结果


# 对5分钟数据进行分析
stats_5min_directional = analyze_price_crossing_directional(df_5min, "5min")  # 分析 5 分钟数据
print("\n5分钟数据 (5min) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印 5 分钟数据分析结果提示信息
print(stats_5min_directional)  # 打印 5 分钟数据分析结果


# 对15分钟数据进行分析
stats_15min_directional = analyze_price_crossing_directional(df_15min, "15min")  # 分析 15 分钟数据
print("\n15分钟数据 (15min) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印 15 分钟数据分析结果提示信息
print(stats_15min_directional)  # 打印 15 分钟数据分析结果
```

```python
import pandas as pd  # 导入 pandas 库，用于数据处理
import plotly.graph_objects as go  # 导入 plotly.graph_objects 库，用于创建图形对象
from plotly.subplots import make_subplots  # 导入 plotly.subplots 库，用于创建子图
import numpy as np  # 导入 numpy 库，用于数值计算

# 加载数据
df_1d = pd.read_csv('2228_data_1d.csv')  # 读取日线数据
df_5min = pd.read_csv('2228_data_5min.csv')  # 读取 5 分钟数据
df_15min = pd.read_csv('2228_data_15min.csv')  # 读取 15 分钟数据

df_1d['date'] = pd.to_datetime(df_1d['date'])  # 将日线数据的日期列转换为 datetime 类型
df_5min['date'] = pd.to_datetime(df_5min['date'])  # 将 5 分钟数据的日期列转换为 datetime 类型
df_15min['date'] = pd.to_datetime(df_15min['date'])  # 将 15 分钟数据的日期列转换为 datetime 类型

print("日线数据 (1d) 示例:")  # 打印日线数据示例
print(df_1d.head())  # 打印日线数据的前几行
print("\n5分钟数据 (5min) 示例:")  # 打印 5 分钟数据示例
print(df_5min.head())  # 打印 5 分钟数据的前几行
print("\n15分钟数据 (15min) 示例:")  # 打印 15 分钟数据示例
print(df_15min.head())  # 打印 15 分钟数据的前几行


def analyze_price_crossing_directional(df, time_frame_name):
    """
    分析价格的向上和向下穿越次数及成交量，并使用 Plotly 绘制分方向的交互式图表。
    左轴显示柱状的穿越成交量（区分向上/下），右轴显示折线穿越次数（区分向上/下）。

    参数:
    df: DataFrame, 输入数据
    time_frame_name: str, 时间周期名称 (例如 "1d", "5min", "15min")
    """

    # 1. 确定价格范围和步长
    min_price = df[['open', 'high', 'low', 'close']].min().min()  # 获取最低价
    max_price = df[['open', 'high', 'low', 'close']].max().max()  # 获取最高价

    start_price = min_price * 0.98  # 设置起始价格，向下扩展 2%
    end_price = max_price * 1.02  # 设置结束价格，向上扩展 2%

    price_range_value = max_price - min_price  # 计算价格范围
    if time_frame_name == '1d':  # 如果是日线数据
        price_step = price_range_value / 50  # 设置价格步长
    elif time_frame_name == '15min':  # 如果是 15 分钟数据
        price_step = price_range_value / 200  # 设置价格步长
    elif time_frame_name == '5min':  # 如果是 5 分钟数据
        price_step = price_range_value / 400  # 设置价格步长
    else:  # 如果是其他时间周期
        price_step = price_range_value / 100  # 设置价格步长
    price_step = max(price_step, 0.01)  # 确保价格步长不小于 0.01

    price_range = []  # 初始化价格范围列表
    current_price = start_price  # 设置当前价格为起始价格
    while current_price <= end_price:  # 循环生成价格范围
        price_range.append(current_price)  # 将当前价格添加到价格范围列表
        current_price += price_step  # 更新当前价格

    directional_crossing_stats = {}  # 存储分方向穿越统计结果的字典

    # 2. 遍历价格，统计分方向穿越次数和穿越成交量
    for price_level in price_range:  # 遍历价格范围
        directional_crossing_stats[price_level] = {'up_crossing_times': 0, 'down_crossing_times': 0, 'up_crossing_volume': 0, 'down_crossing_volume': 0}  # 初始化当前价格水平的分方向统计信息
        daily_data = df.groupby(df['date'].dt.date)  # 按天分组

        for date, daily_df in daily_data:  # 遍历每天的数据
            daily_df = daily_df.sort_values(by='date')  # 确保日内数据按时间排序
            last_close = None  # 初始化上一个收盘价

            for index, row in daily_df.iterrows():  # 遍历日内每条数据
                current_close = row['close']  # 获取当前收盘价
                current_volume = row['volume']  # 获取当前成交量

                if last_close is not None:  # 如果存在上一个收盘价
                    up_crossing = (last_close <= price_level) and (current_close > price_level)  # 判断是否向上穿越
                    down_crossing = (last_close >= price_level) and (current_close < price_level)  # 判断是否向下穿越

                    if up_crossing:  # 如果向上穿越
                        directional_crossing_stats[price_level]['up_crossing_times'] += 1  # 向上穿越次数加 1
                        directional_crossing_stats[price_level]['up_crossing_volume'] += current_volume  # 累加向上穿越成交量
                    elif down_crossing:  # 如果向下穿越
                        directional_crossing_stats[price_level]['down_crossing_times'] += 1  # 向下穿越次数加 1
                        directional_crossing_stats[price_level]['down_crossing_volume'] += current_volume  # 累加向下穿越成交量

                last_close = current_close  # 更新上一个收盘价

    # 5. 结果可视化 (使用 Plotly 绘制分方向的交互式图表 - 使用柱状图和折线图)
    prices = list(directional_crossing_stats.keys())  # 获取价格列表
    up_crossing_counts = [directional_crossing_stats[p]['up_crossing_times'] for p in prices]  # 获取向上穿越次数列表
    down_crossing_counts = [directional_crossing_stats[p]['down_crossing_times'] for p in prices]  # 获取向下穿越次数列表
    up_crossing_volumes = [directional_crossing_stats[p]['up_crossing_volume'] for p in prices]  # 获取向上穿越成交量列表
    down_crossing_volumes = [directional_crossing_stats[p]['down_crossing_volume'] for p in prices]  # 获取向下穿越成交量列表

    fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])  # 创建带次坐标轴的图表

    # 左轴：柱状图 - 向上穿越成交量 (浅绿色)
    fig.add_trace(go.Bar(x=prices, y=up_crossing_volumes, name='向上穿越成交量', marker_color='lightgreen',
                             hovertemplate="价格水平: %{x:.2f}<br>向上穿越成交量: %{y:.2f}",  # 柱状图 hovertemplate
                             ),
                  secondary_y=False, row=1, col=1)  # 浅绿色柱状图

    # 左轴：柱状图 - 向下穿越成交量 (浅红色)
    fig.add_trace(go.Bar(x=prices, y=down_crossing_volumes, name='向下穿越成交量', marker_color='lightcoral',
                             hovertemplate="价格水平: %{x:.2f}<br>向下穿越成交量: %{y:.2f}",  # 柱状图的 hovertemplate
                             ),
                  secondary_y=False, row=1, col=1)  # 浅红色柱状图

    # 右轴：折线图 - 向上穿越次数 (深绿色)
    fig.add_trace(go.Scatter(x=prices, y=up_crossing_counts, name='向上穿越次数', line=dict(color='green'),  # 添加向上穿越次数的折线图
                             hovertemplate="价格水平: %{x:.2f}<br>向上穿越次数: %{y}",  # 设置折线图的悬停提示信息
                             ),
                  secondary_y=True, row=1, col=1)  # 将折线图添加到次坐标轴，深绿色

    # 右轴：折线图 - 向下穿越次数 (深红色)
    fig.add_trace(go.Scatter(x=prices, y=down_crossing_counts, name='向下穿越次数', line=dict(color='red'),  # 添加向下穿越次数的折线图
                             hovertemplate="价格水平: %{x:.2f}<br>向下穿越次数: %{y}",  # 设置折线图的悬停提示信息
                             ),
                  secondary_y=True, row=1, col=1)  # 将折线图添加到次坐标轴，深红色


    # 设置布局
    fig.update_layout(
        title=f'{time_frame_name} - 不同价格水平的向上/下穿越次数和成交量 (柱状图-成交量, 折线图-次数)',  # 设置图表标题，包含时间周期名称
        xaxis_title='价格水平',  # 设置 x 轴标题
        yaxis_title='穿越成交量',  # 设置 y 轴标题（左轴），修改为 "穿越成交量"
        yaxis2_title='穿越次数',  # 设置 y2 轴标题（右轴），保持 "穿越次数"
        yaxis=dict(  # 设置 y 轴（左轴）样式
            title_text="穿越成交量",  # 设置 y 轴标题文本
            title_font=dict(color='black'),  # 设置 y 轴标题字体颜色
            tickfont=dict(color='black'),  # 设置 y 轴刻度字体颜色
            gridcolor='#e0e0e0',  # 设置 y 轴网格线颜色
            zerolinecolor='black',  # 设置 y 轴零线颜色
            tickformat=',.2f',  # 设置 y 轴刻度格式，保留两位小数
            showgrid=True,  # 显示 y 轴网格线
        ),
        yaxis2=dict(  # 设置 y2 轴（右轴）样式
            title_text="穿越次数",  # 设置 y2 轴标题文本
            title_font=dict(color='black'),  # 设置 y2 轴标题字体颜色
            tickfont=dict(color='black'),  # 设置 y2 轴刻度字体颜色
            gridcolor='#e0e0e0',  # 设置 y2 轴网格线颜色
            zerolinecolor='black',  # 设置 y2 轴零线颜色
            tickformat=',.0f',  # 设置 y2 轴刻度格式，整数
            showgrid=True,  # 显示 y2 轴网格线
            overlaying='y',  # 将 y2 轴覆盖在 y 轴上
            side='right'  # 设置 y2 轴在右侧显示
        ),
        xaxis=dict(  # 设置 x 轴样式
            gridcolor='#e0e0e0',  # 设置 x 轴网格线颜色
            zerolinecolor='black',  # 设置 x 轴零线颜色
            tickformat=',.2f',  # 设置 x 轴刻度格式，保留两位小数
            showgrid=True,  # 显示 x 轴网格线
        ),
        plot_bgcolor='#f0f0f0',  # 设置图表背景颜色
        showlegend=True,  # 显示图例
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),  # 设置图例水平显示在图表底部
        hovermode="closest"  # 设置悬停模式为最近点
    )

    # 添加棋盘格背景 (通过修改网格线样式实现类似效果)
    fig.update_xaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True)  # 设置 x 轴次要刻度样式，实现棋盘格效果
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=False)  # 设置 y 轴（左轴）次要刻度样式，实现棋盘格效果
    fig.update_yaxes(minor_ticks="outside", minor_gridcolor='#f0f0f0', minor_griddash='dot', minor_showgrid=True, secondary_y=True)  # 设置 y2 轴（右轴）次要刻度样式，实现棋盘格效果


    fig.show()  # 显示图表
    # [Image of Interactive chart with chessboard grid, refined X-axis ticks and coarser Y-axis ticks showing directional crossing counts (lines) and crossing volume (bars - light color) at different price levels for {time_frame_name} data, with enhanced hover information in Plotly, now with Volume on the left Y-axis and Counts on the right Y-axis]  # 图表描述（注释）


    return directional_crossing_stats  # 返回分方向穿越统计结果


def calculate_optimal_locator_base(data_list):  # 定义计算最优刻度基数的函数
    """
    自适应计算 MultipleLocator 的 base 值， 用于优化刻度密度 (该函数在 Plotly 版本中可能不再直接使用).
    """
    if not data_list:  # 如果数据列表为空
        return 1  # 返回 1
    data_range = max(data_list) - min(data_list)  # 计算数据范围
    if data_range == 0:  # 如果数据范围为 0
        return 1  # 返回 1
    order_of_magnitude = 10**int(np.log10(data_range)) if data_range > 0 else 1  # 计算数量级
    return order_of_magnitude / 5 if order_of_magnitude >= 10 else order_of_magnitude / 2  # 返回最优刻度基数


# 对日线数据进行分析 (函数名修改为 directional 版本)
stats_1d_directional = analyze_price_crossing_directional(df_1d, "1d")  # 分析日线数据
print("\n日线数据 (1d) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印日线数据分析结果提示信息
print(stats_1d_directional)  # 打印日线数据分析结果


# 对5分钟数据进行分析
stats_5min_directional = analyze_price_crossing_directional(df_5min, "5min")  # 分析 5 分钟数据
print("\n5分钟数据 (5min) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印 5 分钟数据分析结果提示信息
print(stats_5min_directional)  # 打印 5 分钟数据分析结果


# 对15分钟数据进行分析
stats_15min_directional = analyze_price_crossing_directional(df_15min, "15min")  # 分析 15 分钟数据
print("\n15分钟数据 (15min) 方向性穿越统计结果 (柱状图-成交量, 折线图-次数):")  # 打印 15 分钟数据分析结果提示信息
print(stats_15min_directional)  # 打印 15 分钟数据分析结果
```
